## HA patterns within a Region

Within a Region, AWS gives you (at least) **three Availability Zones** — physically separate data centres with independent power and network. HA design within a Region is mostly about **using them**.

The standard pattern: **stateless tier across multiple AZs behind a load balancer, stateful tier replicated across AZs, autoscaling on demand.**

- EC2 in an **Auto Scaling Group spanning 3 AZs** behind an ALB.
- **RDS Multi-AZ** with a synchronous standby in another AZ.
- **ElastiCache Redis** with Multi-AZ replication.
- **S3** — the one service where you never think about AZs; it stores objects across them automatically.

Four AWS-native HA primitives worth being explicit about:

- **Multi-AZ Auto Scaling Group** — kills the *single-host* failure mode at zero dev cost. Set `min ≥ 2`, span all AZs, attach a target group, enable cross-zone load balancing.
- **RDS Multi-AZ (synchronous)** — fails the writer endpoint over to the standby on hardware/maintenance failure. **RPO ≈ 0, RTO ≈ 60–120 s**. Single-region only.
- **Aurora storage layer** — data is spread as **six copies across three AZs**; the storage survives an AZ loss transparently, and replicas promote fast.
- **Managed services do AZ work for you** — S3, DynamoDB, SQS, Lambda are multi-AZ by design.

The rule: for production tiers, **span AZs by default** — it's the cheapest resilience you'll ever buy.
