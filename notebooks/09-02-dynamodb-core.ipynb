## DynamoDB — the core model

DynamoDB is a fully managed **NoSQL** store: **tables** hold **items** (records), items hold **attributes** (fields), and items in the same table can have **different attributes** — it's **schemaless apart from the primary key**, which every item must carry. Attribute types are rich (strings, numbers, binary, booleans, null, sets, lists, nested maps); a single item is capped at **400 KB**.

The **primary key** uniquely identifies each item, in two shapes. A **simple key** is just a **partition key (PK)** — each item has a globally unique PK (e.g. `userId`). A **composite key** is **partition key + sort key (PK + SK)** — many items share a PK, and the SK orders them and provides uniqueness (e.g. `customerId` + `orderId`, `deviceId` + `timestamp`).

The core trade is that DynamoDB **rewards designing your access patterns first**, then choosing keys and indexes to support them. The relational habit — normalize the schema, query however you want later — **does not transfer**: you model for the queries you'll run. Pair Lambda with DynamoDB and you get an app that scales from zero to tens of thousands of requests per second with nothing operational to change — which is exactly why it anchors serverless on AWS.
