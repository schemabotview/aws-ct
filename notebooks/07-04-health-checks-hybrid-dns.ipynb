## Route 53 health checks & hybrid DNS

Route 53's **health checkers** are a fleet on the public internet that probe an endpoint from **many locations** and decide healthy/unhealthy by **majority** (typically ~3 of 18+ probers must agree before status flips). Three flavours. **Endpoint** — a direct HTTP/HTTPS/TCP probe on a path, passing on a 2xx/3xx within 4 seconds, with optional **string matching** on the first 5 KB of body (useful when a broken app still returns 200). **Calculated** — a logical AND/OR/NOT of other checks, for "up" that depends on several parts. **CloudWatch alarm** — health read off an alarm; this is the **only way to check a private resource**, since the probers can't reach private IPs (emit a metric → alarm → point the check at it).

Health checks feed routing policies: failover, weighted, latency, and multivalue all drop unhealthy records automatically.

For **hybrid DNS**, every VPC has the built-in **Route 53 Resolver** at base+2. Two endpoint types bridge to on-prem: an **inbound endpoint** lets on-prem servers resolve names in a private hosted zone (on-prem → AWS), and an **outbound endpoint** + **Resolver Rules** forwards queries for, say, `corp.internal` to your on-prem DNS (AWS → on-prem). Each is ENIs (one per AZ for HA), and Resolver Rules can be **shared across accounts via RAM**.
