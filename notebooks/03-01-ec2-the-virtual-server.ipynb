## EC2 — the virtual server

You have an application and you need a computer, reachable from the internet, to run it on — one machine, your code, a URL you can open. That machine is an **EC2 instance**, and EC2 (Elastic Compute Cloud) is AWS's virtual-server service: the most fundamental compute primitive on AWS. Almost every workload runs on EC2 in some form — directly, or as the engine behind a managed service.

An **instance** is a virtual machine. AWS picks the physical hardware, runs a hypervisor on it, and gives you a slice; you choose the OS, CPU, memory, storage, and networking, and AWS handles everything underneath.

Getting from nothing to a server you can open in a browser is **six decisions**: an **AMI** (the disk image to boot), an **instance type** (size), a **key pair** (to log in), the **network** (a VPC + a public IP), a **security group** (the firewall — open ports 80 and 22), and **user data** (a boot script that turns a bare OS into a running app). Make the happy-path choice for each and you have your "it works" moment.

The rest of this module revisits each of those six decisions with judgment, then scales the pattern out: **one instance → a load balancer in front of many → an Auto Scaling Group that grows and shrinks the fleet automatically.**
