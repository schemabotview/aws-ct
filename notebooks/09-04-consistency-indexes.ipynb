## Read consistency & secondary indexes

Writes commit to multiple AZs before acknowledging; **reads come in three consistency levels**. **Eventually consistent** (default) may be ~1 s stale, costs **0.5 RCU** — fine for most reads. **Strongly consistent** always returns the latest committed write, costs **1 RCU** — for read-your-own-write. **Transactional** gives **ACID across multiple items/tables**, costs 2× again. Pattern: eventually-consistent by default, strong only where the logic needs it, transactions for the small set of true multi-item invariants (financial, inventory).

**Secondary indexes** let one table answer more access patterns by keeping a second copy keyed differently. A **Local Secondary Index (LSI)** keeps the base **PK** but a **different sort key** ("same customer, by date") — must be defined **at table creation**, up to **5**, shares the table's throughput, supports **strong reads**, and caps each PK at 10 GB. A **Global Secondary Index (GSI)** uses a **different partition key** (± sort key) ("all orders where status=SHIPPED, by date") — can be **created/dropped anytime**, up to **20**, has **its own throughput**, and is **eventually consistent only**.

The rule: **GSI is the default** (flexible, add anytime); **LSI** only when you need strong consistency on an alternate sort within the same partition — and remember you must plan LSIs up front.
