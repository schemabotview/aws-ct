## AWS Backup

**AWS Backup** is the managed, central backup service covering **EC2, EBS, RDS, Aurora, DynamoDB, EFS, FSx, S3, Storage Gateway, Neptune, DocumentDB, and Redshift** — the long list is the point. Without it you'd run a different backup configuration per service.

The model has three pieces:

- **Backup vaults** hold the recovery points. Each vault has a **KMS key** and an access policy. **Vault Lock** makes it **WORM** — once locked in compliance mode, even the **root user cannot** shorten retention or delete recovery points. This is the **ransomware-resilience** answer.
- **Backup plans** define **schedules, retention, and lifecycle** transitions to cold storage. One plan can target many resource types.
- **Resource assignment** matches resources to plans **by tag** (or explicit ARN). Tag `backup=daily-prod` and the plan picks them up automatically.

Two features worth highlighting:

- **Cross-Region copy** in the backup plan automatically duplicates recovery points to another Region — the **cheapest DR** building block (this *is* Backup & Restore, automated).
- **Cross-account copy** to an isolated **backup account** protects against a compromised production account deleting its own backups.

The exam cue: *"one place to back up many service types,"* *"immutable backups against ransomware,"* *"tag-based backup policy across the org"* → **AWS Backup** (with Vault Lock, cross-Region/-account copy).
