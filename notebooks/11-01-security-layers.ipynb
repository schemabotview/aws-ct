## The six security layers at a glance

Security on AWS is **layered** — no single service does the whole job; each owns one layer, and a defensible stack pulls from most of them.

- **KMS** is the cryptographic root. It creates and protects the keys that encrypt data across every other service — S3, EBS, RDS, DynamoDB, Secrets Manager all call it underneath.
- **Secrets Manager** (and its cheaper cousin **SSM Parameter Store**) holds credentials so applications never embed passwords or API keys in code.
- **Shield** absorbs DDoS at the network edge; **WAF** filters HTTP traffic at the application layer; **Firewall Manager** centralises both across an Organisation.
- **Cognito** handles user identity — who is signed in, and what AWS resources they may reach.
- A handful of **detective services** — GuardDuty, Macie, Inspector, Detective — watch for misuse after the fact, with **Security Hub** aggregating their findings.

The mental split is **preventive** (encrypt, hide credentials, keep bad traffic out, prove identity) versus **detective** (notice when something slipped through). Everything in this module maps to one of those two jobs.
