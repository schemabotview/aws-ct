## VPC peering

**VPC Peering** is a direct, **point-to-point** connection between two VPCs that lets instances reach each other by **private IP**, as if on the same network — **same or cross account, same or cross region**, over AWS's backbone with no bandwidth cap.

The constraints are sharper than expected. **CIDRs cannot overlap** — if both VPCs use `10.0.0.0/16`, peering won't establish. Peering is **non-transitive**: if A peers with B and B peers with C, A still **cannot** reach C through B — you need a direct A↔C peering. And after accepting the request you must **update route tables on both sides** to point at the peering connection, or the link exists but no traffic flows.

A nice property: within the same region, an SG rule can **reference a security group in the peered VPC** — "allow 5432 from the app-tier SG in the other VPC" works directly. Peering is the right tool for **two to a few VPCs**; past a handful, the full mesh (N×(N−1)/2 connections) collapses and you move to Transit Gateway.
