## Multi-AZ vs. read replicas

The two RDS features that look alike and solve **completely different problems**. **Multi-AZ** provisions a **synchronous standby** in another AZ in the same region — every write commits to **both** before the app is acknowledged. The standby carries **no traffic** (a hot spare); when the primary fails, RDS promotes it and **flips the endpoint's DNS**, so the app reconnects to the *same* endpoint name — typical failover **60–120 s**. Its whole purpose is **high availability, not capacity**. The newer **Multi-AZ DB Cluster** runs one writer + two *readable* standbys across three AZs — HA plus a little read scaling, and failover often **under 35 s**.

**Read Replicas** are **asynchronous** copies that **do serve read traffic** — up to 5 per source (Aurora: 15). Each has its **own endpoint**; the app must be written to send reads there. Replicas can be **cross-region** (also a DR option) and can be **promoted** to a standalone primary. Because replication is async, replicas can lag — fine for reporting/analytics, wrong for read-after-write consistency.

The one-line split: **Multi-AZ = availability (synchronous, idle standby, same endpoint); Read Replicas = read scaling (asynchronous, active, separate endpoint).** They're often used **together** — Multi-AZ for HA, replicas for read fan-out.
