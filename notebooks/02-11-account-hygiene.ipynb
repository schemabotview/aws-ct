## Account hygiene — Access Analyzer & MFA

Once the mechanics work, two tools keep an account from drifting. The first is **IAM Access Analyzer**, which watches for unintended openness and for permissions creeping away from least privilege. It does four things: **external-access** findings continuously inspect resource policies (S3, IAM roles, KMS, Lambda, SQS, Secrets Manager) and flag anything reachable from outside your account or org; **unused-access** findings flag IAM users, roles, permissions, and access keys not used lately; **policy validation** checks a policy for syntax and best-practice problems *before* you deploy it; and **policy generation** reads CloudTrail history for an identity and proposes a tighter policy from what it actually used.

The second is **Multi-Factor Authentication (MFA)** — a second factor layered over a password, so a stolen password alone isn't enough to get in. Three factor types: a **virtual authenticator app** (Google Authenticator, Authy, 1Password), a **hardware TOTP token**, or a **FIDO security key** (YubiKey and similar).

The simple rule: **MFA on root, MFA on every IAM user with console access, and MFA *enforced* through condition keys** on the operations that hurt most when they go wrong — `iam:*`, `s3:DeleteObject`, and the like. Together, Access Analyzer tells you where you've drifted and MFA raises the cost of a stolen credential — the two habits that keep a mature account healthy.
