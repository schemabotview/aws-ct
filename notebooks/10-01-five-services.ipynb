## Five services, one question

How do components of a distributed system pass work to each other **without being tightly coupled**? A direct API call from producer to consumer ties their fates: a slow consumer slows the producer, a down consumer fails it, a second consumer means changing the producer. Five services answer different versions of that question — all getting producers and consumers **off each other's hot path**, letting each side scale independently, surviving either's failure, and adding retries and observability for free.

**SQS** is a **queue** — producers drop messages in, one consumer pulls each out at its own pace. **SNS** is **pub/sub** — one publish, every subscriber gets a copy. **Kinesis** and **MSK** are real-time **streams** for high-volume, multi-consumer, replayable event data. **Step Functions** is **workflow orchestration** — a managed state machine sequencing tasks with branches and retries. **EventBridge** is an **event bus** routing AWS-service, SaaS, and custom events to targets **by rule**.

The difference is **shape**: one-to-one queue vs. one-to-many fan-out vs. persistent stream vs. orchestrated workflow vs. rule-routed bus. The module walks the messaging pair (SQS/SNS + the triangle with EventBridge), then streaming (Kinesis/Firehose/MSK), then Step Functions, then EventBridge — and closes with a decision sheet.
