## Global Accelerator

**Global Accelerator** is the third leg of the edge story. It **caches nothing** — instead it gets user traffic onto **AWS's private backbone** at the nearest edge and carries it across AWS infrastructure to the app, giving less variability, fewer hops, and more predictable latency than the public-internet path.

The user-facing surface is **two static Anycast IPv4 addresses** (or bring your own). **Anycast** means the same IP is announced from many edges, so a user lands at the closest one; from there AWS routes over the backbone to the healthy, lowest-latency endpoint. **Two static IPs that never change** is the power: clients hardcode them, firewalls allowlist them, and **failover happens at the network layer** — no DNS TTL to wait on.

It routes to **ALBs, NLBs, EC2, and Elastic IPs**, organized into **endpoint groups** (one per region), each with a **traffic dial** (0–100%) to control its share — set to 0 to **drain a region** during a deploy — and **endpoint weights** + built-in health checks within a group. **Client IP preservation** keeps the original client IP for ALB/EC2 endpoints. The whole point: **static IPs, second-scale failover, and a private backbone path for any TCP/UDP workload** — gaming, VoIP, IoT, non-HTTP APIs.
