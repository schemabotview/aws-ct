## Regions — and how to choose one

A **Region** is a physical geographic area holding a cluster of AWS data centers — Northern Virginia, Ireland, Frankfurt, Mumbai. Each has a code like `us-east-1`, `eu-west-1`, `ap-south-1`; the number is just an identifier, not a count.

**The key property is independence.** Regions don't share power, networking, or control planes, so a failure in one has no causal link to another. Most resources live *in* a region — an S3 bucket, an RDS database, a VPC — and to have a presence elsewhere you deploy there explicitly. The exceptions are **global services** — IAM, Route 53, CloudFront, Organizations — where the region selector grays out. Note the **`us-east-1` quirk**: new services launch there first, and several global control planes call it home, so its outages can have a wider blast radius.

**Choosing a region** comes down to four factors, in rough order of weight:

1. **Compliance & data residency** — regulations may force data to stay in a country or economic zone. Often non-negotiable.
2. **Latency** — the closest region to your users wins; spread-out users call for multi-region behind Route 53 / CloudFront.
3. **Service availability** — verify every service you need exists in the target region.
4. **Pricing** — varies modestly; a tiebreaker, not a driver.
