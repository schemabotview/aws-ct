## The three identity types — users, groups, roles

IAM gives you three kinds of identity, and the difference between them is mostly about **credentials**.

A **user** is a single human or a long-lived application. It has permanent credentials — a console password, or an access key and secret. The rule is **one user per person**; never share a user between people.

A **group** is just a bag of users — a way to attach one set of policies to many people at once. Groups don't authenticate (there are no group credentials), they can't nest, and a user can belong to many.

A **role** is an identity with policies but **no permanent credentials**. Nobody logs in *as* a role — something *assumes* it, and at that moment AWS vends **short-lived credentials that auto-expire**. Roles are the most important of the three, because they're how almost every credential problem is solved cleanly:

- an **EC2 instance** talks to S3 (instance profile),
- a **Lambda function** reads DynamoDB (execution role),
- **account A** reaches a bucket in **account B** (cross-account role),
- **federated employees** enter AWS at all (SAML / OIDC).

**Rule of thumb:** if you're about to create a long-lived access key to solve a problem, the answer is almost always a **role** instead.
