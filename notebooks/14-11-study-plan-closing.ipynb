## Study plan & closing

A practical four-week run at SAA-C03 once you've read this series:

- **Week 1 — Foundations, IAM, VPC, Compute** (modules 01, 02, 03, 06). Build a lab: a VPC with public + private subnets, an ASG behind an ALB across two AZs, an IAM role with least-privilege policies. Practise reading IAM policy JSON.
- **Week 2 — Serverless, Storage, DNS, CDN** (04, 05, 07). Deploy Lambda + API Gateway; an S3 bucket with versioning + lifecycle + a CloudFront distribution; Route 53 with a failover routing policy.
- **Week 3 — Data, Integration, Security** (08, 09, 10, 11). Build an RDS Multi-AZ instance; a DynamoDB table with GSIs; an SQS-to-Lambda pipeline; encrypt an EBS volume with a KMS CMK.
- **Week 4 — Observability, Resilience, Cost** (12, 13). Run an AWS Backup plan; wire CloudWatch alarms; do a Well-Architected review. Then **timed practice exams** until you consistently clear the bar, reviewing every wrong answer to the underlying principle.

### Closing

Fourteen modules in, the shape of AWS should be steady: an **account** is a security + billing boundary; **Regions and AZs** are the physical substrate; **IAM** enforces identity at the chokepoint; a portfolio of **compute / storage / network / data / integration / security** services each solves a specific problem at a specific layer; and the **operational disciplines** — observability, deployment, backup, DR, cost — turn the parts into a running system.

AWS is wide but it **composes.** Every workload you design from here uses the same primitives in different combinations. The work is no longer learning what the pieces are — it's **recognising which combination fits the requirement in front of you.** The Well-Architected Framework is the vocabulary for the trade-offs; the SAA exam is the calibrated test of whether you can name the right combination under constraints. You have the map. Go build.
