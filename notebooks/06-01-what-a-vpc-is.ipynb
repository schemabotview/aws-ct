## What a VPC is

A **VPC** — Virtual Private Cloud — is your own **logically isolated network** inside AWS. Every resource that touches the network lives in one: EC2 instances, RDS databases, VPC-attached Lambda functions, ECS tasks, load balancers. The VPC defines an **IP address space**, how that space is sliced into **subnets**, what traffic can **enter or leave**, and how to reach the rest of the world — other VPCs, AWS services, on-prem data centres, the public internet.

The module splits into three problems. **Inside one VPC**: the fixed set of decisions you make standing up a fresh network — the CIDR block, subnets across AZs, route tables, what's allowed in and out (security groups and NACLs), how a private instance reaches the internet (NAT), and how you get a shell on something with no public IP. **Connecting VPCs and AWS services privately**: peering, endpoints, PrivateLink, and Transit Gateway. **Reaching on-premises**: Site-to-Site VPN and Direct Connect.

Everything downstream is one of those three: build the network, connect it to other networks and services, or bridge it to your data centre — then a decision table for picking the right tool.
