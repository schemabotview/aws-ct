## The relational data tier on AWS

Almost every application needs a **relational database** — Postgres, MySQL, SQL Server — the durable, transactional **system of record**. You could run the engine yourself on EC2 (install, patch, replicate, script backups, build failover) and own all that operational surface, or hand it to AWS and just bring the schema and queries. This module is the three services that make up the data tier.

**RDS** is AWS's managed relational service — you bring the schema and queries; AWS handles the OS, engine binary, patching, backups, and failover. **Aurora** is AWS's own cloud-native engine that speaks the **MySQL and PostgreSQL wire protocols** but rebuilt the storage and replication layer from scratch — a different performance ceiling and availability story. **ElastiCache** is the third piece — managed **in-memory caching** (Redis, Valkey, or Memcached) that sits **in front of** the database to absorb hot reads and turn millisecond queries into microsecond cache hits.

The module walks RDS (engines, backups, availability, security, RDS Proxy), then Aurora (its storage architecture, Serverless v2, Global Database, and the RDS-vs-Aurora decision), then ElastiCache (why cache, the patterns, Redis vs Memcached) — and closes on how the three **stack** into one tier: durable source of truth + connection pooling + a cache on the read path.
