## RDS vs. Aurora — picking which

Both are managed relational databases; the choice is **engine, performance ceiling, and how variable the workload is**.

- **RDS (Oracle / SQL Server / Db2)** — when the engine **isn't on Aurora's list** (Aurora only speaks MySQL/PostgreSQL).
- **RDS MySQL / PostgreSQL** — cost-sensitive, predictable workload, fine with **60–120 s failover**.
- **Aurora MySQL / PostgreSQL** — want **faster failover**, faster replication, up to **15 replicas**.
- **Aurora Serverless v2** — variable workload, multi-tenant SaaS, dev/test, unknown scale.
- **Aurora Global Database** — cross-region DR with **RPO < 1 s**, or global low-latency reads.
- **Aurora Backtrack** — guard against accidental wipes on Aurora MySQL.

**Cost shape matters.** Aurora is priced per-second compute + per-GB-month storage + per-million IO. RDS is per-second instance + **pre-provisioned EBS**. For very small steady workloads RDS is often cheaper; for anything bursty or growing, Aurora's storage model usually wins. The default heuristic: **MySQL/PostgreSQL → Aurora unless a small steady workload makes RDS cheaper; anything else → RDS.**
