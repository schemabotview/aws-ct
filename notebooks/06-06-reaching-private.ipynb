## Reaching private instances — EIPs, VPC DNS, Flow Logs

Three supporting pieces round out one VPC. An **Elastic IP** is a static IPv4 you own at the account level and attach to any ENI — it **survives instance stops** (a regular public IP is released on stop) and **remaps** to another instance in minutes (the failover pattern). Free while attached to a running instance, a small hourly charge when idle, 5 per region by default. Use them where external clients need a **fixed IP** (NAT Gateway, VPN endpoint, on-prem allowlist) — not on app servers; that's what load balancers are for.

**VPC DNS** has two toggles: **enableDnsSupport** turns on the AWS resolver (at VPC base + 2) that resolves AWS endpoints and Route 53 private zones; **enableDnsHostnames** gives public-IP instances public DNS names. Both on in the default VPC, off in custom — flip both on to use Route 53 **private hosted zones**. The **Route 53 Resolver** endpoints forward queries to/from on-prem DNS for hybrid networks.

**VPC Flow Logs** capture packet **metadata** (src/dst IP, port, protocol, bytes, **ACCEPT/REJECT**) — not contents — to S3, CloudWatch, or Firehose. They're the ground truth when traffic misbehaves: a silently-dropping SG or NACL shows up as **REJECT** entries pinpointing the rule.

Finally, **getting a shell** on a private instance: the old way was a **bastion host** (a hardened public EC2 jump box); the modern default is **SSM Session Manager** — a shell to any SSM-agent instance with **no open SSH port, no public subnet, no key pair**, fully audited. **EC2 Instance Connect** is the SSH-style equivalent. New infra defaults to Session Manager; bastions are the fallback.
