## AWS Config — what changed, and is it compliant

**AWS Config** records the **configuration of every supported resource** as it changes, stores that history in S3, and evaluates each resource against **rules**. The output is a per-resource timeline — *"what did this security group look like at 14:32 last Tuesday, and was it compliant?"*

The flow:

1. Config **records** the configuration of every supported type — EC2, S3, RDS, IAM, security groups, VPCs — each time something changes.
2. **Configuration history** lands in your S3 bucket; **snapshots** capture point-in-time state.
3. Config **evaluates** each resource against the rules attached to your account.
4. **Notifications** fire on compliance-state changes via SNS or EventBridge.
5. Optional **remediation** auto-fixes non-compliant resources via **SSM Automation**.

### Config Rules

| Type | What it is | Examples |
|---|---|---|
| **AWS managed** | 200+ pre-built rules | `encrypted-volumes`, `s3-bucket-public-read-prohibited`, `restricted-ssh` |
| **Custom** | a Lambda you write | any org-specific policy |

**Conformance packs** bundle rules + remediation into a deployable YAML template, so a whole compliance baseline (PCI, CIS) ships as one unit — and via **Organizations** across every account.

The distinction from CloudTrail: CloudTrail says *who made the change*; Config says *what the resource looks like now, what it looked like before, and whether that's compliant*. They pair.
