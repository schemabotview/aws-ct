## Compliance — Artifact, Audit Manager, programmes

Day-to-day observability is one thing; **compliance** is periodic audits, certifications, and evidence packages. Two services handle the paperwork, and they answer different questions.

### AWS Artifact — *AWS's* compliance posture

The self-service portal for AWS's **own** compliance documentation — no support ticket needed. Two pieces:

- **Reports** — downloadable: **SOC 1/2/3**, **PCI DSS** Attestation, **ISO 27001**, **FedRAMP** packages, GDPR docs, pen-test summaries. Hand these to *your* auditors as evidence of AWS's posture.
- **Agreements** — legal documents you accept: the **HIPAA Business Associate Agreement (BAA)**, the **GDPR Data Processing Addendum**. Scoped to one account or a whole Organisation.

Cue: *download AWS's SOC 2*, *sign the BAA for HIPAA* → **Artifact**.

### AWS Audit Manager — *your* audit evidence

Continuously **collects evidence** from your environment and maps it to framework controls (SOC 2, PCI, HIPAA, GDPR). It pulls in **CloudTrail events, Config snapshots, and Security Hub findings**, and assembles an audit-ready report — so you're not screenshotting consoles the week before an audit.

The split: **Artifact = AWS proves *its* compliance to you; Audit Manager = you prove *your* compliance to an auditor.** Config and CloudTrail are the evidence sources underneath.
