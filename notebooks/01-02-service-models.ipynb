## The three service models — IaaS, PaaS, SaaS

The service model is the question of **how much of the operational stack you still own** after you rent from AWS. Three levels, each handing more to the provider.

**IaaS — Infrastructure as a Service.** AWS manages the virtualization, networking, and hardware; you manage the operating system, runtime, application, and data. This is the most control and the most work — **EC2, EBS, VPC**.

**PaaS — Platform as a Service.** AWS also manages the OS and runtime; you bring only the application and its data. Less to run, less to tune — **Elastic Beanstalk, RDS**.

**SaaS — Software as a Service.** AWS manages the entire stack; you just use the application and own your data inside it — **WorkMail, Chime, QuickSight**.

Each step up the stack trades **flexibility for less operational burden**. There's no "best" level — you pick per workload: raw EC2 when you need control, a managed platform when you'd rather not run the plumbing.
