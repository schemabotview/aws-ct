## Picking the right edge service

All three services sit "at the edge" but answer different questions — sort by **layer**: DNS resolution, application-layer caching, or network-layer backbone routing.

- **Resolve a hostname; control which endpoint by weight/geo/latency/health** → **Route 53**
- **Cache HTTP/HTTPS content close to users; static sites, media, high cache-hit** → **CloudFront**
- **Get non-HTTP (TCP/UDP) onto the backbone fast, static IPs, near-instant failover** → **Global Accelerator**

Two sharper splits. **CloudFront vs. Global Accelerator**: CloudFront caches and is HTTP-only with DNS-speed (minutes) failover and changing IPs; Global Accelerator doesn't cache, handles any TCP/UDP, fails over at network-speed (seconds), and gives static IPs — the standard answer for gaming, VoIP, IoT. **Route 53 latency routing vs. Global Accelerator**: both send users to the lowest-latency region, but Route 53 does it via **DNS** (user then rides the public internet) while Global Accelerator does it via **Anycast at the network layer** (user lands on the AWS backbone at the nearest edge, failing over faster).

Framed by *DNS vs. cache vs. backbone*, and *HTTP vs. any protocol*, the right edge service is usually obvious.
