## CloudTrail — who did what

**CloudTrail** records **every API call** in your account — console clicks, CLI invocations, SDK calls, service-to-service calls. Every event captures the same five fields:

- **Who** — the IAM principal (user, role, federated identity, or AWS service)
- **What** — the API action: `RunInstances`, `PutObject`, `CreateBucket`, `DeleteTrail`
- **When** — timestamp
- **Where** — source IP, user agent, region
- **Result** — success or error code, plus request/response parameters

It's the audit trail behind every *"who did that?"* question and the foundation of incident investigation.

### Event types

| Type | Captures | Default | Cost |
|---|---|---|---|
| **Management** | control-plane: create / modify / delete, IAM, network | **on, free** | first copy free |
| **Data** | data-plane: S3 `GetObject`, Lambda `Invoke`, DynamoDB reads | opt-in | charged |
| **Insights** | ML-flagged anomalies in API volume (spike in `TerminateInstances`) | opt-in | charged |

### Getting it right

- A **multi-region trail** to an **S3 bucket** captures every region in one place — the standard baseline.
- **Log file validation** + **S3 Object Lock** detect and prevent tampering — critical for compliance, since an attacker's first move is often `DeleteTrail`.
- Deliver a trail to **CloudWatch Logs** to alarm on sensitive actions in near-real-time (e.g. root login, `DeleteTrail`).
