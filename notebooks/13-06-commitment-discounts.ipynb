## Commitment discounts — Reserved Instances, Savings Plans, Spot

For anything running steadily, paying On-Demand leaves **30–70% on the table**. AWS offers three families of discount:

- **Reserved Instances (RIs)** — commit to a specific **instance family + region** for **1 or 3 years**; up to **~72% off**. **Standard RIs** (deepest discount, locked to the exact family) vs. **Convertible RIs** (smaller discount, exchangeable for other families mid-term). Best for predictable steady-state on one family.
- **Savings Plans (SPs)** — commit to an hourly **dollar amount** of spend; the discount applies **across families and regions** automatically. **Compute SP** covers EC2, Fargate, and Lambda interchangeably (most flexible); **EC2 Instance SP** is family-pinned but slightly deeper. Easier than RIs for fleets that churn.
- **Spot Instances** — bid on unused EC2 capacity, up to **~90% off**. AWS can reclaim with a **2-minute notice**. Perfect for batch jobs, CI runners, stateless workers, and Fargate Spot — never for a stateful database.

The standard layering:

1. **Baseline** — the always-on floor → **Savings Plans** (or RIs).
2. **Variable steady load** → additional Savings Plans as it stabilises.
3. **Burst / fault-tolerant** → **Spot**.
4. **Spiky, unpredictable** → **On-Demand** for the tip.

The rule: **commit on the base, burst on Spot, pay On-Demand only for the unpredictable tip.** RIs vs. SPs cue: *need cross-family flexibility* → Savings Plans; *a specific reserved capacity guarantee in an AZ* → zonal RIs.
