## Why cloud — traditional IT vs. cloud

You want to run software for other people to use. You have two options: buy servers, rack them in a building you cool and secure, and hire people to look after them — or **rent that capacity by the hour** from a company that already does all of that. **Cloud is the second option, treated as the default.** Almost everything downstream — what you give up, what you get — falls out of that one choice.

**Cloud computing** is the *on-demand delivery of compute, storage, networking, and dozens of other services over the internet*, paid for by what you actually consume. The provider owns the hardware, the buildings, the cooling, the security, and the network. You request resources through an API; they appear in seconds; you pay only while they exist.

That shift — from **owning** capacity to **renting** it — changes almost every architectural decision. Against traditional IT: hardware moves from *buy-and-own* to *rent on-demand*; capacity from *over-provision for peak* to *scale up or down in minutes*; provisioning from *weeks* to *seconds*; and the cost model from a large upfront **CapEx** to pay-as-you-go **OpEx**. The racks, power, and cooling — the *undifferentiated heavy lifting* — become the provider's problem, not your team's.

That removes the classic pain points: paying for idle capacity, multi-week procurement, guessing your capacity needs eighteen months out, and running a physical data center at all. What you give up is direct control of the metal; what you get is **speed, elasticity, and global reach in minutes**.
