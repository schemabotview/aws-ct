## Cost tools — Cost Explorer, Budgets, Anomaly Detection

Three tools, three jobs.

- **Cost Explorer** — the **interactive analysis** console. Slice spend by service, account, region, or tag over arbitrary time windows. The **Recommendations** tab proposes RIs and Savings Plans from the last 7/30/60 days; the **Forecast** view projects month-end and next-month spend. Where finance and engineering meet — same data, two questions. *(This is analysis, not alerting.)*
- **AWS Budgets** — the **proactive alarms**. Set a monthly **cost** budget, a **usage** budget (e.g. "RI coverage > 80%"), or a **Savings Plans utilisation** budget, and Budgets emails or fires **SNS** at thresholds (50 / 80 / 100% and *forecasted* overage). The trick: budget **per team or per workload**, not just the account total — a one-team overrun is invisible inside a healthy whole-account budget.
- **Cost Anomaly Detection** — the **ML** layer. Learns each service's normal spend pattern and alerts on unexpected spikes. The classic catch: a cryptominer on a leaked IAM key spinning up GPU instances — the bill spikes hours before anyone notices the breach.

And the enablers underneath: **cost allocation tags** (activate them so Cost Explorer/Budgets can slice by team/project) and **AWS Cost and Usage Report (CUR)** — the exhaustive line-item export to S3 for deep analysis in Athena/QuickSight.

Cue map: *analyse/slice historical spend* → Cost Explorer · *alert before crossing a threshold* → Budgets · *catch an unexpected spike* → Anomaly Detection.
