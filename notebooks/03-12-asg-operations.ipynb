## ASG operations — refresh, hooks, warm pools

Three operational features make an ASG production-grade. First, **health checks**. The default **EC2** check only sees host-level failure; the **ELB** check also fails an instance that's *running but returning 500s*. **Always enable ELB health checks on production ASGs** — otherwise a broken-but-running instance keeps serving traffic. A **health check grace period** delays checks on new instances so they can boot and pass first.

Second, **Instance Refresh**. When the launch template changes (new AMI, new user data), Instance Refresh rolls the change across the fleet automatically — capped by a **minimum healthy percentage** so only a slice is out at once, with a **warm-up time** before a new instance counts as in-service, and optional checkpoints to pause and verify. It's the engine of **immutable infrastructure**: don't patch live boxes — bake a new AMI, point the template at it, refresh.

Third, **lifecycle hooks** and **warm pools** for slow starts. A **lifecycle hook** pauses an instance in `Pending:Wait` (or `Terminating:Wait`) so you can bootstrap, run smoke tests, or drain connections before it joins or leaves — up to a timeout. A **warm pool** keeps pre-initialised instances (stopped or hibernated) ready to activate in **seconds** — pair it with a lifecycle hook so the slow warm-up work happens **once, before** the instance enters the pool.
