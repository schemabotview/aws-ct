## CloudWatch Metrics

A **metric** is a time-series of numeric data points — CPU percent, request count, error rate, latency. Metrics are the foundation of CloudWatch; alarms and dashboards are built on top of them.

Every metric has:

- **Namespace** — a logical grouping. AWS services emit into `AWS/EC2`, `AWS/RDS`, `AWS/Lambda`; custom metrics use any namespace you choose.
- **Metric name** — `CPUUtilization`, `Invocations`, `Duration`, `OrdersProcessed`.
- **Dimensions** — key/value pairs identifying the resource: `{InstanceId: i-abc123}`. Each unique combination is a **separate metric stream**.
- **Resolution** — standard is **1-minute** granularity; high-resolution custom metrics go down to **1 second**.

### Standard vs. custom

AWS services push **standard metrics for free** — EC2 CPU / network / disk I/O, S3 bucket size, Lambda invocations and duration, ALB request counts.

The catch worth memorising on EC2: standard metrics come from the **hypervisor**, so **memory utilisation and in-OS disk space are NOT collected by default**. Those need the **CloudWatch Agent** (next-but-one section).

**Custom metrics** are anything you push via `PutMetricData` — application counters, business KPIs. Use dimensions to slice them, high resolution when one-second granularity matters (e.g. fast auto-scaling reactions).
