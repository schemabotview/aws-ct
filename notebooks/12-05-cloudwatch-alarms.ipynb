## CloudWatch Alarms

An **alarm** watches a single metric (or a metric-math expression) and transitions between three states:

| State | Meaning |
|---|---|
| **OK** | within threshold |
| **ALARM** | breached, per the evaluation rules |
| **INSUFFICIENT_DATA** | not enough data — startup, gaps, paused service |

### Evaluation parameters (these trip people up)

A 5-minute **period**, 3 **evaluation periods**, and **2 out of 3 datapoints to alarm** means: *ALARM when the metric breached in at least 2 of the most recent 3 five-minute windows.* Set them too tight and the alarm **flaps**; too loose and response is delayed.

### Actions

- **SNS notification** — the universal output: email, SMS, Lambda, fan-out anywhere.
- **Auto Scaling action** — scale an ASG in/out (the target-tracking / step-scaling backend).
- **EC2 action** — stop, terminate, reboot, or **recover** (recover replaces the underlying host on hardware failure, preserving instance ID + EBS volumes).

### Two variants

- **Composite alarm** — combines several alarms with AND/OR logic, so you page only when *multiple* conditions hold — kills alert-storm flapping.
- **Billing alarm** — an alarm on the `EstimatedCharges` metric (only in **us-east-1**) to watch the AWS bill.
