## S3 access & features — pre-signed URLs, access points, events, hosting

A **pre-signed URL** grants **temporary, credential-free** access to one specific object — the signature carries the authorization, inheriting whatever the signer can do, and expires in at most 7 days (STS-signed URLs are bounded by the session). The two big uses: let a user download a private object via a link, or let a browser **upload directly to S3** without round-tripping your backend.

**S3 Access Points** are named endpoints on a bucket, each with its **own policy** — instead of one tangled bucket policy serving every team, you make `analytics-ap` (read-only), `app-ap` (write path), `vpc-ap` (VPC-only), each focused. **Event Notifications** fire when objects are created, removed, restored, or transitioned, routed to **SQS, SNS, Lambda, or EventBridge** — direct destinations for one-trigger-one-handler, **EventBridge** once routing gets complex (pattern matching, multiple targets, replay).

For **static website hosting**, S3 can serve a site directly (index + error docs on the `s3-website` endpoint), but production puts **CloudFront** in front: custom domain, **HTTPS via ACM**, edge caching, and **Origin Access Control (OAC)** so the bucket stays private and only CloudFront can read it — with WAF at the edge. Bare endpoint for quick internal sites; CloudFront + OAC is the production default.
