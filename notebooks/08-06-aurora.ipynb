## Aurora — a different engine & endpoints

**Aurora** isn't managed open-source MySQL/PostgreSQL — it's AWS's own engine that speaks their wire protocols on the front but **reimplements the storage and replication layer** underneath. That rebuild is where its numbers come from: **~5× the throughput of RDS MySQL, ~3× of RDS PostgreSQL** at the same instance type. The mechanism is the **shared cluster volume** — instead of each instance owning its own data files and replication meaning "stream write events to replay elsewhere," Aurora **separates compute from storage**: primary and replicas share one distributed, log-structured **cluster volume** that keeps **six copies across three AZs** (two per AZ), with writes acknowledging on a **4-of-6 quorum**. Replicas read the same volume, so replication lag is tiny and adding a replica adds no storage.

An Aurora cluster exposes several **endpoints**, each routing differently: the **cluster (writer) endpoint** → the current primary (writes); the **reader endpoint** → load-balanced across all replicas (reads); a **custom endpoint** → a subset you define (e.g. big instances for reports); an **instance endpoint** → one specific instance (debugging). If the primary fails and a replica is promoted, **writer and reader endpoints update automatically** — the app never has to know which instance is currently the writer, and the reader endpoint is a DNS-level read load balancer you don't manage.
