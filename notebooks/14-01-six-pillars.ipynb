## The Well-Architected Framework — six pillars

Thirteen modules in, you have the AWS service map. This one ties the threads together. AWS's **Well-Architected Framework** is the structured way the field answers *"is this architecture good?"* — **six pillars**, each with design principles, best-practice questions, and a portal-based scoring tool.

| Pillar | The question it answers |
|---|---|
| **Operational Excellence** | Can we run and improve the system day to day? |
| **Security** | Are data and identities protected against credible threats? |
| **Reliability** | Do we meet our uptime and recovery targets? |
| **Performance Efficiency** | Do we hit performance targets without overspending? |
| **Cost Optimization** | Are we paying only for value delivered? |
| **Sustainability** | Are we minimising environmental impact? |

Every service you've already met fits under one or more pillars — this module is **composition, not new introductions**.

The pillars are **not independent**. Tightening security often costs more; improving reliability usually trades against cost; sustainability often argues the same way as cost. The Well-Architected **review process exists to make those trade-offs explicit per workload**, so nobody silently optimises one pillar at another's expense. That tension — naming the trade-off and choosing deliberately — *is* the job of a solutions architect, and the through-line of this final module.
