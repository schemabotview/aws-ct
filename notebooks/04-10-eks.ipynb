## Amazon EKS — managed Kubernetes

ECS covers the vast majority of new container workloads. Three situations push you to **EKS** instead: an **existing Kubernetes investment** (one operational model across environments), **multi-cloud or on-prem portability** (Kubernetes manifests are largely cloud-agnostic; ECS task definitions aren't), and **rich Kubernetes-specific needs** (taints, tolerations, custom controllers, Helm, service meshes, GitOps).

**EKS** runs the Kubernetes **control plane** — API server, etcd, scheduler, controller manager — as a managed service (~$0.10/hour per cluster). You own the **worker nodes**, via three options: **Managed Node Groups** (AWS handles the EC2 lifecycle — the sensible default), **Self-Managed Nodes** (your AMI, your ASG — for custom AMIs, Windows, kernel needs), or a **Fargate Profile** (no EC2 at all — pods matching a namespace/label selector run on Fargate). EKS Fargate has real trade-offs: no DaemonSets, no privileged containers, no EBS-backed volumes (EFS works), no GPUs.

The choice comes down to Kubernetes being a **real requirement, not a preference**. Default to **ECS on Fargate** for new AWS-only workloads — simplest, cheapest, lowest overhead. Reach for **EKS** when you already run Kubernetes, need portability, or need its scheduler/ecosystem; the control-plane cost and operational complexity earn their place then.
