## CloudWatch Logs

Logs are the **unstructured cousin of metrics** — text lines from services and applications, collected and queryable. The hierarchy:

- **Log Group** — the container; one per logical source, with a **retention** setting (1 day to 10 years, or never expire).
- **Log Stream** — events from a single emitter (one Lambda execution environment, one EC2 instance, one container).
- **Log Events** — the actual timestamped lines.

Lambda writes automatically to `/aws/lambda/<function>`. ECS captures container stdout/stderr; API Gateway has access logs; VPC Flow Logs, Route 53 query logs, CloudTrail, and the CloudWatch Agent all land here too.

### Metric filters

A **metric filter** scans incoming log events for a pattern and increments a CloudWatch metric on a match. The canonical pattern: a filter like `?ERROR ?Exception` on an app log group emits an `ApplicationErrors` metric, and an **alarm on that metric pages you**. This is how you turn unstructured logs into actionable signals.

### Logs Insights & subscriptions

- **CloudWatch Logs Insights** — a SQL-like query language to search across log groups ad-hoc (`fields`, `filter`, `stats … by`) — the tool for "grep production logs without SSH."
- **Subscription filter** — streams matching log events in **real time** to OpenSearch, Lambda, or Kinesis (Data Streams / Firehose) for downstream processing or a third-party SIEM like Splunk.
