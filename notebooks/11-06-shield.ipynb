## Shield — DDoS protection

Shield protects AWS resources from **distributed denial-of-service** attacks. Two tiers.

### Shield Standard

Automatic, **free**, and on for every account. It defends **Layer 3 and Layer 4** — SYN floods, UDP reflection, volumetric attacks — for resources behind CloudFront, Route 53, Elastic Load Balancers, and EC2 Elastic IPs. Nothing to configure; AWS absorbs routine DDoS traffic before it reaches your infrastructure.

### Shield Advanced

A paid subscription at **$3,000 per organisation per month** (plus data transfer). It builds on Standard with the things worth memorising:

- **Layer 7 DDoS protection** — defends application-layer floods (HTTP GET/POST floods, slowloris) against CloudFront, ALB, and API Gateway. Standard covers only L3/L4.
- **DDoS cost protection** — AWS credits the scaling charges (EC2, ELB, CloudFront, Route 53) you incur **because of** an active attack. Without it, a successful absorption can still arrive with a six-figure bill.
- **24/7 Shield Response Team (SRT)** — direct access to AWS DDoS experts during an attack, and the ability to pre-authorise them to act on your account.
- **Advanced real-time metrics and reporting**, plus **global threat dashboards**.

The exam cues: *free, automatic, L3/L4* → Shield Standard. *L7 protection, cost protection, the SRT, or a bill after an attack* → Shield Advanced. Shield defends **availability**; WAF (next) inspects **content** — they layer together at the edge.
