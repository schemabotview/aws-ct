## DNS & record types

DNS resolves a hostname to an address. The browser asks a **recursive resolver** (its ISP's, or a public one like `1.1.1.1`), which walks the hierarchy — root → TLD → the domain's authoritative name servers. **Route 53 is the authoritative tier** for any hosted zone it manages. The resolver caches the answer for the record's **TTL** and re-asks when it expires — so **TTL is your failover lever**: a low TTL (say 60 s) propagates changes fast; a high TTL means fewer queries (cheaper) but slower cutover.

The record types to know: **A** (host → IPv4), **AAAA** (host → IPv6), **CNAME** (host → another hostname), **NS** (zone → its authoritative name servers), **MX** (mail routing, with priority), **TXT** (arbitrary text — domain verification, SPF, DKIM), **SRV** (service location with priority + port).

A **hosted zone** is the container for a domain's records. **Public** zones answer the internet; **private** zones answer only from associated VPCs (so `db.internal` → a private IP), and need the VPC's `enableDnsSupport` + `enableDnsHostnames` on. Route 53 is also a **registrar** — register there and it auto-creates the zone and wires NS records; register elsewhere and you just paste Route 53's four NS records into that registrar.
