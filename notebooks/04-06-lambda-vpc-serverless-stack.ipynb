## Lambda permissions, VPC & the serverless stack

Lambda has **two IAM sides**. The **execution role** controls *what the function can do* — Lambda assumes it at runtime, so the credentials in your handler are role credentials (apply least privilege). The **resource-based policy** controls *who can invoke the function* — it's added automatically when you wire a trigger in the console (API Gateway, S3), and set manually for cross-account invocation.

By default Lambda runs **outside your VPC**, in an AWS-managed network with internet access. To reach **private** resources (RDS, ElastiCache, an internal service) you attach it to a VPC subnet — but then it **loses internet** unless you route through a **NAT Gateway**, exactly like a private EC2. Modern **Hyperplane ENIs** make the VPC cold-start overhead negligible, so today it's a reachability decision, not a latency one.

Lambda needs an HTTP front, and it's rarely alone. A **Function URL** is a built-in HTTPS endpoint (one per function — webhooks, prototypes); **API Gateway** is the production front door, adding routing, throttling, API keys, caching, custom domains, WAF, and Cognito/Lambda authorizers. The wider **serverless stack** pairs Lambda with **DynamoDB** (NoSQL that scales to zero), **S3**, **SQS/SNS/EventBridge** (decoupling), **Step Functions** (orchestration), **Aurora Serverless**, and **Cognito** — each scaling independently and billing per use.
