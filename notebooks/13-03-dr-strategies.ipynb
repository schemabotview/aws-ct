## DR strategies — the four AWS shapes

Where HA handles everyday failures within a Region, **DR** handles the rare, catastrophic ones — a whole Region unreachable, ransomware, a bad change that deletes production. AWS frames DR as **four named strategies**, ordered by RTO/RPO (and cost). Pick per workload — most enterprises run two or three across their portfolio.

| Strategy | RPO | RTO | Idea |
|---|---|---|---|
| **Backup & Restore** | hours | hours–days | Periodic backups to S3 / AWS Backup. On disaster, provision new infra in the DR Region and restore. **Cheapest, slowest.** |
| **Pilot Light** | minutes | tens of min | Core **data** continuously replicated to the DR Region; compute kept off/minimal. On disaster, scale up + switch DNS. |
| **Warm Standby** | seconds–min | minutes | Full but **scaled-down** stack always running in the DR Region, kept current. On disaster, scale up + shift traffic. |
| **Multi-Site Active-Active** | seconds | seconds | Both Regions serve traffic; data replicated bidirectionally. On disaster, traffic shifts automatically. **Costliest, fastest.** |

The progression is also a progression in **how much runs warm** in the DR Region: nothing → data only → a small live stack → a full second site. The cost climbs with it.

The building blocks under these shapes: **Aurora Global Database** and **DynamoDB Global Tables** (cross-region data replication), **S3 Cross-Region Replication**, and **Route 53 health-check failover** (the DNS switch that shifts traffic when the primary fails). Whatever the shape, **rehearse it** — an untested DR plan is a hope, not a plan; run failover drills quarterly.
