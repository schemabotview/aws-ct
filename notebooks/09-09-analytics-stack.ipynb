## The analytics stack — Athena, Glue, EMR, OpenSearch

Four services complete the analytics picture around Redshift, all sitting on **S3 as the data lake**. **Athena** is **serverless SQL over S3** — no infrastructure, no loading; point it at S3 objects (via the Glue Data Catalog), write SQL, pay **per TB scanned**. It's for **data-lake querying** when a full warehouse is overkill: log analysis, ad-hoc queries, dashboards over Parquet. Athena vs. Redshift Spectrum look alike; the split is that Spectrum is part of a Redshift cluster (cluster pricing) while Athena is standalone — **use Athena when you don't have Redshift, Spectrum when you do**.

**Glue** is managed **ETL + the Glue Data Catalog** (the metastore for S3 tables). **Crawlers** scan S3 and infer schemas into the catalog, which **Athena, Spectrum, and EMR all read**; **Glue ETL** jobs are serverless Spark billed per DPU-second. The catalog is the shared source of truth for "what tables exist in the lake."

**EMR** is managed **Hadoop/Spark** — the full open-source big-data ecosystem (Spark, Hive, Presto, HBase, Flink) on a managed cluster, for custom or heavy processing that Glue's serverless Spark doesn't fit. **OpenSearch** is managed **full-text search + log analytics** (the Elasticsearch fork) with OpenSearch Dashboards — for search, observability, and log exploration; a very common pairing is **DynamoDB Streams → OpenSearch** to add search over operational data.
