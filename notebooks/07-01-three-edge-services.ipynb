## Three edge services, one job

Every request starts with a **DNS lookup** — the browser turns a hostname into an IP before it connects — and the bytes then have to travel from the user to your origin and back over a fixed speed of light. AWS's three **edge** services each attack a different layer of that path, all answering some version of *"get the user to the right endpoint, fast."*

**Route 53** is AWS's authoritative **DNS** — it answers name lookups for domains you host and decides *which endpoint* a request lands on, via routing policies and health checks. **CloudFront** is the **CDN** — hundreds of edge locations that **cache** content close to users, absorb load off the origin, and terminate TLS at the edge. **Global Accelerator** is the third leg — it **caches nothing**; instead it gets **non-HTTP (TCP/UDP)** traffic onto **AWS's private backbone** at the nearest edge and carries it across AWS infrastructure, with **static IPs** and network-speed failover.

Three services, three layers of the user-to-app path: **DNS resolution** (Route 53), **application-layer caching** (CloudFront), and **network-layer backbone routing** (Global Accelerator). The module walks each, then a decision matrix.
