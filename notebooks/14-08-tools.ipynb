## The tools — Well-Architected Tool & Trusted Advisor

The pillars give you the principles; two AWS services turn them into practice you can **score against**.

### AWS Well-Architected Tool

Lives in the console. You define a **workload** (name, Regions, lens), pick **lenses** — the AWS Well-Architected Framework lens by default, plus specialised lenses for **Serverless, SaaS, Foundational Technical Review (FTR), Machine Learning, IoT, Data Analytics, Hybrid Networking** — and walk the questionnaire.

Each question's unchecked options become **HRIs (High Risk Issues)** or **MRIs (Medium Risk Issues)**, scored per pillar. You add improvement notes, link tickets, and **re-run every 6–12 months** to watch the workload's posture trend. **Custom lenses** encode organisation-specific standards on top of the AWS baseline — useful in regulated industries.

### Trusted Advisor

The always-on companion. It runs checks across five categories — **Cost Optimization, Performance, Security, Fault Tolerance, Service Limits** — into a coloured dashboard. The **basic tier** (free, all accounts) gives seven core checks (chiefly security: root MFA, public S3, exposed keys, IAM use — plus Service Limits); **Business / Enterprise Support** unlock the full hundreds. It can publish to **EventBridge** — *"alert me when a new public S3 bucket appears"* is one rule.

Use them together: **Well-Architected for periodic workload-level reviews; Trusted Advisor for account-level continuous monitoring.**
