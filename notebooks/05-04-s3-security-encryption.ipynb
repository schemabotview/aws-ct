## S3 security & encryption

Every bucket and object is **private by default**; access is granted through layered controls. **Block Public Access (BPA)** is the safety net — on by default for new buckets, it **overrides any policy** that would make objects public and runs *before* anything else (a bucket policy granting `Principal: "*"` is silently ignored if BPA is on). Below it: **bucket policies** (resource-based, the primary cross-account/conditional grant), **IAM policies** (same-account principals), and legacy **ACLs** (disabled by default since 2023 — leave them off). The rule that catches people: an **explicit deny anywhere always wins**.

Since 2023, every new object is **encrypted by default** with **SSE-S3** (AWS-managed keys). You override for stronger guarantees: **SSE-KMS** uses your KMS key — every decryption is a **CloudTrail-logged** API call, giving a second access lever independent of the bucket policy — the right answer for regulated data. **SSE-C** means you send the key with each request; **client-side** means AWS only ever sees ciphertext.

SSE-KMS's catch is cost and throughput (KMS calls are billable and rate-limited); **S3 Bucket Keys** cache a bucket-level data key and cut KMS calls by ~99%. Enforce HTTPS-only with an `aws:SecureTransport` condition.
