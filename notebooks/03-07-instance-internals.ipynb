## Instance internals — Nitro, tenancy, placement

Under every modern instance sits the **Nitro System** — AWS's platform that **offloads** virtualization (networking, storage, security) from the host CPU onto dedicated Nitro cards and runs a stripped-down hypervisor. The payoff: **near bare-metal performance** (most host CPU and RAM go to your instance), hardware EBS encryption at no CPU cost, network ceilings up to 100–200 Gbps, and a hard **security boundary** where AWS operators have no path into guest memory. Nitro also enables **bare metal** (`*.metal`) instances with no hypervisor at all — rarely needed, except for nested hypervisors, socket-based licensing, or measuring raw hardware behaviour.

**Tenancy** controls how much hardware you share. **Shared** (the default) puts you on hardware with other customers, isolated at the hypervisor — cheapest, and the right choice almost always. A **Dedicated Instance** reserves hardware for your account; a **Dedicated Host** gives you a specific physical server with visible sockets — the one to pick when a licence is tied to physical sockets or an audit needs host-level visibility.

**Placement groups** override AWS's default spreading: **cluster** packs instances close for low-latency HPC, **spread** puts each on distinct hardware for resilience, **partition** groups them into fault-isolated racks. And the **stop / hibernate / terminate** trio differ on state: stop keeps the EBS root (RAM lost), **hibernate saves RAM to the encrypted root** so a slow-to-warm service resumes instantly, and terminate deletes the instance and (by default) its root volume.
