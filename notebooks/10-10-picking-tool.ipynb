## Picking the right tool

The whole module in one decision sheet — match the need to the service.

- **Decouple producer/consumer, buffer work, one consumer per message** → **SQS Standard**
- **...with strict order + exactly-once** → **SQS FIFO**
- **Isolate failing messages without blocking the queue** → **SQS DLQ**
- **One event → many subscribers at once** → **SNS** (with ordering → SNS FIFO)
- **Durable fan-out to many independent workers** → **SNS + SQS**
- **Real-time stream, many consumers, replay history** → **Kinesis Data Streams**
- **Bulk-deliver streaming data to S3 / Redshift / OpenSearch** → **Firehose**
- **SQL/Flink over streams (windows, joins, anomalies)** → **Managed Flink**
- **Kafka semantics or existing Kafka clients** → **MSK**
- **Multi-step workflow — branches, retries, parallel, audit** → **Step Functions Standard**
- **Very high-throughput short workflow (≤ 5 min)** → **Step Functions Express**
- **Pause for human approval / external callback** → **`.waitForTaskToken`**
- **Route AWS/SaaS/custom events by rule; scheduled cron jobs** → **EventBridge** (+ Scheduler)

The meta-pattern: **queue vs. fan-out vs. stream vs. workflow vs. bus** — and real systems **combine** them (SNS+SQS fan-out, Kinesis→Firehose→S3, EventBridge Pipes into Step Functions). Match the *shape* of the coupling, not the service name.
