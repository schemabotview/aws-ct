## EventBridge — the event bus

**EventBridge** is a serverless **event bus** that routes events to targets by **rules** — where SNS does pub/sub by *subscription*, EventBridge does routing by *pattern match*; the **rule** is the abstraction. Three bus kinds: the **default bus** (every account has one; AWS services auto-publish service events here — EC2 state changes, S3 events, CodePipeline, IAM activity, hundreds more), a **custom bus** (your app's events via `PutEvents`), and a **partner bus** (SaaS partners like Datadog, Stripe, GitHub publish into your account). An event is JSON with standard fields (`source`, `detail-type`, `account`, `region`, `time`) and a free-form `detail` — the same shape across AWS/SaaS/custom.

A **rule** has an **event pattern** (filters incoming events with rich syntax — exact, lists, prefix, numeric ranges, anything-but, existence) and up to **five targets** matching events go to. Targets can be almost anything: Lambda, Step Functions, SQS, SNS, Kinesis, another bus.

Three features seen in real architectures: **Schema Registry** auto-discovers event schemas and generates typed **code bindings** (Python/Java/TS) as the event surface grows; **Pipes** is point-to-point **source → filter → enrich → target** with no glue code (replaces the "Lambda that polls one service and forwards" pattern, filtering at the source); and **Archive + Replay** stores matching events for a retention period and replays them (debugging, reprocessing after a bug fix). EventBridge also runs **Scheduler** (cron/rate jobs).
