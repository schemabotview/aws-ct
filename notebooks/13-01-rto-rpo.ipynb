## RTO and RPO — the only two numbers that matter

Every production workload eventually answers two questions: *can we survive a failure?* (high availability) and *how fast can we come back?* (disaster recovery). The whole conversation reduces to two numbers, quoted **per workload**:

- **RTO — Recovery Time Objective** — how long the workload can be **down** before the business is in real trouble.
- **RPO — Recovery Point Objective** — how much **data** the business can afford to lose, measured in time (back to the last good backup or replication point).

Quote them per workload, never per company. The checkout flow might need **RTO < 5 min, RPO < 30 s**; the back-office reporting warehouse is fine with **RTO 24 h, RPO 4 h**. Every architectural choice that follows — synchronous vs. asynchronous replication, Multi-AZ vs. multi-region, hot standby vs. cold backup — is **paying for a smaller RTO/RPO**.

Two costs scale as you tighten them:

- **Money** — more hardware, more replicas, more cross-region bandwidth.
- **Complexity** — more failover automation, more primary/standby drift, more to test.

The only honest question is *how small is small enough for this workload?* — and this module (HA, DR, cost, migration) is the set of levers you pull to hit those two numbers **without overpaying**.
