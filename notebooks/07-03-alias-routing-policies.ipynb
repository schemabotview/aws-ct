## Alias records & routing policies

An **Alias record** is a Route 53 extension that points a hostname directly at an **AWS resource** rather than another hostname. To the client it looks like an A record returning an IP; to you, no hard-coded IP to break. It beats a CNAME in four ways: it's **allowed at the zone apex** (a CNAME isn't), **free to query**, has **AWS-managed TTL**, and **integrates with health checks**. Valid targets to memorise: **ELBs, CloudFront, API Gateway, S3 website endpoints, Elastic Beanstalk, VPC interface endpoints, Global Accelerator**, and same-zone records. **EC2 and RDS endpoints are *not* aliasable** — those still need CNAMEs (so can't sit at the apex).

**Routing policies** decide *which value* Route 53 returns when records share a name. **Simple** — one record, no health awareness. **Weighted** — proportional split by weight (0–255), health-aware — the canary/A-B tool. **Latency** — returns the region with the lowest measured latency from the resolver. **Failover** — active/passive, health-gated to a standby. **Geolocation** — by the user's continent/country (compliance, localized content). **Geoproximity** — by geographic distance with an adjustable bias. **Multivalue answer** — several health-checked records, client picks — a poor-man's load balancer. **IP-based** — route by the client's CIDR.

Most non-trivial designs **combine two or three** (e.g. latency to pick a region, then weighted for a canary within it).
