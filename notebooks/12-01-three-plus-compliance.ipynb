## Three observability services, plus a compliance layer

Three services answer three different questions about a running AWS environment — and one model underpins all of them.

- **CloudWatch** answers *how is it performing?* — metrics, logs, alarms, dashboards.
- **CloudTrail** answers *who did what?* — every API call across the account, recorded and queryable.
- **AWS Config** answers *what changed, and is it compliant?* — a timeline of every resource's configuration, with rules that fire on drift.

Underneath the three sits the **Shared Responsibility Model** — the line between what AWS secures and what the customer configures. On top of them sits the **compliance layer**: **AWS Artifact** (self-serve portal for AWS's own audit reports) and **Audit Manager** (evidence collection for *your* audits).

A one-line mnemonic carries the whole module: **CloudWatch = how is it doing, CloudTrail = who did what, Config = what changed and is it compliant.** Everything else is a variant or a comparison against those three.
