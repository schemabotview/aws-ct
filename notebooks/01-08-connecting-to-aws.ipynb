## How you talk to AWS

Every interaction with AWS goes through the **same control plane** — there are just three doors into it, and underneath they all do the same thing.

- **AWS Management Console** — a web UI in the browser. Best for learning a new service and one-off exploration.
- **AWS CLI** — commands from your terminal. Best for scripting, automation glue, and ad-hoc operations.
- **SDKs** (like `boto3` for Python) — programmatic access from your own code. Best for application logic, infrastructure-as-code, and repeatable workflows.

Underneath all three are **signed HTTPS requests to a regional API endpoint**. The console is a web app calling those endpoints; the CLI is a thin wrapper over them; the SDK is a typed wrapper in your language. Learn a service in the console, then move to the CLI or an SDK once you want it **scripted and repeatable**.
