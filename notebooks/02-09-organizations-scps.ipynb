## AWS Organizations & SCPs

Most real companies don't have one AWS account — they have dozens or hundreds, split by environment, team, blast radius, and billing line. **AWS Organizations** turns that fleet into one managed tree: a **root** at the top, nestable **Organizational Units (OUs)** as containers, and **accounts** as the leaves. Two account types matter: the **management account** owns the org, pays all bills, and is the only one that can attach SCPs or create members (**don't run workloads there**); every other account is a **member**.

You reach for Organizations for four things: **consolidated billing** (one invoice, volume discounts and Savings Plans across the fleet), **programmatic account creation** (new accounts as code — the basis of landing zones), **centralized governance** (rules enforced from above), and **cross-account sharing** (RIs, Savings Plans, AMIs via RAM).

The governance lever is the **Service Control Policy (SCP)** — attached to the root, an OU, or an account, it defines the **maximum permissions** for every principal in scope, *including member-account root*. Two things to keep straight: an SCP **does not grant** anything — it only caps, and IAM must still grant the action; and it **doesn't apply to the management account** or to service-linked roles. Most orgs run a **deny-list** (start from `FullAWSAccess`, add explicit `Deny`s). Two patterns to recognize: a **region restriction** (`Deny` when `aws:RequestedRegion` isn't approved, with `NotAction` exempting global services so they don't break) and an **org lock-in** (`Deny` on `organizations:LeaveOrganization`).
