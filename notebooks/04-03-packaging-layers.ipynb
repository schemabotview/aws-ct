## Packaging, layers & configuration

You ship code into Lambda three ways, by size: **inline zip** in the console (up to 3 MB source, for tiny utilities), a **zip via S3** (250 MB unzipped, the standard path), or a **container image** (OCI, up to 10 GB — for large dependencies, ML models, or custom runtimes). **Lambda Layers** are versioned zip archives of shared libraries or runtimes that mount into the environment — up to five per function, shared across functions and accounts — so you keep packages small and avoid re-uploading big dependencies on every deploy.

Configuration is a handful of knobs. **Memory** runs 128 MB to 10,240 MB — and **CPU scales with it** (one full vCPU around 1,769 MB). **Timeout** is 1 second to 15 minutes, a hard cap that kills the function if exceeded. **Ephemeral `/tmp`** is 512 MB to 10 GB and survives across warm invocations. **Environment variables** cap at 4 KB, encrypted at rest with KMS. And **account concurrency** defaults to 1,000 per region (a soft limit).

The single most important sizing knob is the **memory–CPU link**: there is no direct CPU dial, so for CPU-bound work you **raise memory to get more vCPU**, even if you don't need the extra RAM. Sizing memory is really sizing CPU.
