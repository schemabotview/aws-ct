## The database zoo — why purpose-built

You've met relational on RDS and Aurora — extraordinarily **general** (almost any data shape fits rows and columns; SQL expresses almost any query). So why does AWS publish a whole **zoo** of other databases? Because at scale, **generality has a cost**: a relational engine doing petabyte analytics, microsecond key-value lookups, billion-edge graph traversal, or time-series sensor data is slower and pricier than a purpose-built one. Each service here answers **one access pattern** brilliantly — and almost everything else badly. The skill is **cue → service**.

The two anchors: **DynamoDB** is AWS's serverless **NoSQL** (key-value + document) — single-digit-millisecond reads/writes at any scale, no servers, no upfront schema, auto-replicated across 3 AZs; it underpins most serverless architectures. **Redshift** is AWS's **data warehouse** — columnar, massively-parallel, for analytical queries over TB–PB.

Around them sit **purpose-built** databases (Neptune graph, DocumentDB, Keyspaces, Timestream, MemoryDB, QLDB) and an **analytics stack on S3** (Athena, Glue, EMR, OpenSearch). This module walks DynamoDB in depth, then Redshift + the analytics stack, then the rest of the zoo, and closes with a cue-to-service cheat sheet.
