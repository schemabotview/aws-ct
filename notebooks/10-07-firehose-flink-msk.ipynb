## Firehose, stream analytics & MSK

**Kinesis Data Firehose** is the managed **delivery** sibling of Data Streams — a **one-way pipe**: point it at a destination, hand it data, and it buffers, batches, and delivers, with **no shards, no consumers, no checkpointing**. The trade-off: it **stores nothing** (failed records can back up to S3, but you **can't replay history**). Destinations: **S3** (most common, pairs with Athena/Glue), **Redshift** (writes to S3 then `COPY`s), **OpenSearch**, HTTP endpoints, and third-parties (Splunk, Datadog…). It batches by **buffer size (1–128 MB)** or **interval**, and can run a **Lambda transform** (and Parquet/ORC conversion) inline. **Streams vs. Firehose: Streams when you build consumers and need replay; Firehose when you just want data delivered to a destination.**

**Managed Service for Apache Flink** (formerly Kinesis Data Analytics) runs **real-time analytics over streams** in Flink (Java/Scala/Python) or SQL: windowed aggregations (tumbling/sliding/session), stream joins, anomaly detection, writing results back to a stream or Firehose — fully managed, no cluster. For per-minute dashboards, threshold alerts, on-stream ETL.

**Amazon MSK** is fully managed **Apache Kafka** — topics, partitions, brokers, consumer groups — with AWS handling provisioning, patching, and multi-AZ, so existing Kafka clients/connectors work unchanged. **MSK provisioned** (you size brokers) or **MSK Serverless** (auto-scales). **Kinesis vs. MSK: Kinesis if you're AWS-native and want zero ops; MSK if you already run Kafka or need its ecosystem.**
