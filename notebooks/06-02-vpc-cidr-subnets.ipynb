## VPC, CIDR & subnets

A VPC is **region-scoped** — it spans every AZ in a region — while **subnets are AZ-scoped**. You pick an IPv4 **CIDR** at creation, from `/16` (65,536 addresses) down to `/28` (16), and can add **secondary CIDRs** later if you outgrow it. The planning rules: **CIDRs must not overlap** with anything you'll peer with (use `10.0.0.0/16`, `10.1.0.0/16`, `10.2.0.0/16` for distinct environments, not the same block thrice); the **RFC 1918** private ranges are conventional but any range routes. Every account gets a **default VPC** per region (`172.31.0.0/16`, a public subnet per AZ, IGW attached) — fine for experiments, wrong for production.

A **subnet** is a slice of the VPC CIDR in one AZ. Crucially, **"public" and "private" aren't properties of the subnet** — they're a consequence of its **route table**: a subnet is public because its route table has a route to an Internet Gateway, private because it doesn't. Re-associate the route table and its identity flips.

One gotcha: AWS **reserves 5 IPs per subnet** (the first four + the last), so a `/24` gives **251 usable**, not 256. This bites when you size small subnets for a Lambda attachment or interface endpoints and come up short.
