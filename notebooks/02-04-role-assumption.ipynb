## Role assumption — three policies, two sides

Assuming a role is a **two-sided handshake**, and it turns on **three policies** that all have to agree. Two live on the role; one lives on the caller.

On the **role** there are two. A **trust policy** answers *who is allowed to assume this role* — the EC2 service, account `222222222222`, a specific IAM user. A **permissions policy** answers *what the role can do once assumed* — read this S3 bucket. On the **caller** there is a third: an **identity policy** granting `sts:AssumeRole` on that role — permission to *make the call at all*.

All three must align. If the caller's identity policy is missing, AWS returns `AccessDenied` on the `AssumeRole` call itself — before the trust policy is even read. The trust policy is necessary but not sufficient.

Concretely, a backup bot assuming an `S3BackupRole`: the **bot** carries only `Allow sts:AssumeRole` on the role — *nothing about S3*. The role's **trust policy** names the bot as an allowed principal. The role's **permissions policy** carries the actual `s3:GetObject / PutObject`. The bot has no S3 power of its own; it only has the right to *become* the role. That's the whole point — broad credentials exist only as short-lived sessions, never as long-lived keys on the bot. **Same-account and cross-account look identical** — cross-account just means the trust policy names a principal in another account.
