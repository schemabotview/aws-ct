## Amazon ECS — object model & task definitions

**ECS** is AWS's native container orchestrator — it schedules, places, scales, and replaces containers across a cluster. Its object model nests: a **Cluster** (a logical fleet, backed by EC2 or Fargate) holds **Services**, each maintaining a desired number of **Tasks**, and a Task is a running instance of a **Task Definition**. The **task** is the unit of deployment (one or more containers together); the **service** is the unit of long-running operation (it holds the desired count, integrates with a load balancer, and replaces failed tasks). One-off batch jobs run as standalone tasks.

A **task definition** is a JSON blueprint pinning down everything about how a task runs: image, CPU/memory, port mappings, env vars, log driver, volumes, network mode — and **two distinct IAM roles**. The **task execution role** is used by the **ECS agent on the host** — plumbing to pull the image from ECR and ship logs (`ecr:BatchGetImage`, `logs:PutLogEvents`). The **task role** is your **application's identity** — whatever your app needs (`s3:GetObject`, `dynamodb:Query`).

Splitting them is deliberate: the execution role is infrastructure plumbing; the task role is your app. Credentials reach the container via the standard SDK metadata path, so `boto3` inside the container just works with no embedded keys.
