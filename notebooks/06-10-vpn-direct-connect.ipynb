## Reaching on-prem — Site-to-Site VPN & Direct Connect

Two ways bridge AWS to your data centre. A **Site-to-Site VPN** runs an **IPsec encrypted tunnel** over the **public internet** — fastest to stand up (minutes, no install), encrypted by default, but limited in throughput and subject to internet jitter. Its three parts: a **Virtual Private Gateway (VGW)** on the VPC (or a TGW VPN attachment for multi-VPC), a **Customer Gateway (CGW)** representing your on-prem device's public IP, and the **VPN Connection** running the tunnels. Every connection has **two tunnels** to two AWS endpoints for automatic failover. Routing is **static** or **dynamic via BGP** (the default for non-trivial topologies; required for **VPN CloudHub**, the multi-branch pattern).

**Direct Connect (DX)** is a **dedicated physical fibre** from your facility to an AWS Direct Connect location — bypassing the public internet for **consistent bandwidth, predictable latency**, and a compliant private path, but it takes **weeks** to provision. One DX carries **Virtual Interfaces (VIFs)**: a **Private VIF** (VPC resources), a **Public VIF** (AWS public endpoints over the link), and a **Transit VIF** (a TGW via a DX Gateway). A **DX Gateway** reaches VPCs in **multiple regions** from one connection.

Production hybrid almost always runs **both — DX as primary, VPN as same-day fallback** (DX takes weeks, so the VPN bridges the gap).
