## AMIs & bootstrapping

An **AMI** (Amazon Machine Image) is the template an instance launches from — the OS, pre-installed software, and baked-in config. The AMI itself is just metadata; the actual bytes live in EBS snapshots AWS manages for you. AMIs come from four sources: **AWS-provided** (Amazon Linux, Ubuntu, Windows — patched by AWS), **Marketplace** (vendor images with licensing), **Community** (public, shared by other accounts), and **Custom** (your own golden image). **AMIs are region-scoped** — the same logical image has a different ID per region, and you `CopyImage` to use it elsewhere.

The deeper choice is **how much you bake in versus bootstrap at boot**. **Fully bootstrapped** does all setup in user data — slowest to launch, most flexible, good for dev. **Fully baked** puts the app and config into the AMI — fastest launch, least flexible, the right call for **Auto Scaling at scale and immutable infrastructure**. Most production fleets sit in the **hybrid** middle: a base AMI with dependencies, app pulled at launch. **EC2 Image Builder** is the AWS-native pipeline for producing golden AMIs on a schedule or when the base image updates.

One gotcha the exam likes: **deregistering an AMI does not delete its snapshots** — they keep billing until you remove them explicitly. And sharing an **encrypted** AMI means also sharing the **KMS key**, or the recipient can't launch it.
