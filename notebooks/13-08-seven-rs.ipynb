## Migration — the 7 Rs

Most AWS customers don't arrive greenfield. They arrive with a data centre, a portfolio of workloads, and a deadline. **Migration** is moving from there to here without breaking the business in transit. The framing AWS pushes is the **7 Rs** — every workload is being:

| R | What it means | When |
|---|---|---|
| **Retire** | turn it off | unused or duplicated |
| **Retain** | leave on-prem (for now) | latency, compliance, or lock-in make the move uneconomic |
| **Rehost** | lift-and-shift to EC2 | fastest move, minimal change — *"forklift to the cloud"* |
| **Relocate** | move VMware → VMware Cloud on AWS | specifically the VMware path |
| **Replatform** | lift-and-tinker (e.g. DB → RDS) | modest gains, no app rewrite |
| **Repurchase** | replace with SaaS (drop the CRM, buy Salesforce) | a SaaS equivalent exists |
| **Refactor** | rebuild cloud-native (monolith → microservices, serverless) | highest value, highest effort |

Picking the R is half the migration; the service that moves the bytes is the other half.

The practical sequencing most enterprises follow: **Retire and Retain first** (cut the scope you don't move), then **Rehost the majority** to get out of the data centre fast, then **Replatform and Refactor over time** as workloads earn the investment. Trying to Refactor everything up front is the classic way to miss the deadline — get to the cloud first, modernise once you're there.
