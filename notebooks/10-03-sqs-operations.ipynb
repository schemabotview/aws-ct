## SQS operations — DLQs, polling, delays, large messages

Four SQS features come up constantly. A **Dead-Letter Queue (DLQ)** is a separate queue for messages that fail too many times: set `maxReceiveCount` (commonly 3–5) on the main queue, and after that many failed deliveries SQS moves the message aside (the DLQ must match type — FIFO source → FIFO DLQ). Two payoffs: **poison messages** (malformed payloads that break the consumer every time) get shunted aside so the rest of the queue keeps moving, and the DLQ is the **audit log of failures** — inspect it, fix the bug, and **redrive** the messages back into the source.

**Long polling** — set `WaitTimeSeconds` 1–20 on `ReceiveMessage` so SQS waits for a message instead of returning empty immediately (short polling). It **cuts empty receives, cost, and latency** and should be the default. **Delay queues** postpone a message's visibility for up to **15 minutes** after send (queue-wide default, or per-message with **message timers**). For payloads over **256 KB**, the **Extended Client Library** stores the body in **S3** and passes a reference through the queue.

Together these are the operational toolkit: **DLQ** to isolate failures, **long polling** for efficiency, **delays** for scheduling, and the **S3 extended client** for big payloads.
