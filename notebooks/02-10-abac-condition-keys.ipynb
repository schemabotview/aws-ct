## ABAC — tag-based access & condition keys

Conditions are what make policies **context-aware** — the optional `Condition` block that checks the situation before allowing an action. A short list of condition keys solves most real problems: **`aws:MultiFactorAuthPresent`** (force MFA before sensitive actions, paired with `Deny`), **`aws:SourceIp`** (restrict to office / VPN ranges), **`aws:RequestedRegion`** (region restriction), **`aws:PrincipalOrgID`** (only principals from your org), **`aws:SecureTransport`** (require HTTPS), and **`sts:ExternalId`** / **`aws:SourceArn`** for confused-deputy protection.

Two of those keys — **`aws:PrincipalTag`** and **`aws:ResourceTag`** — are the building blocks of **ABAC: Attribute-Based Access Control.** Instead of hard-coding ARNs into policies, ABAC writes rules whose effect depends on **tags**. Tag your principals with `Project=Atlas`, tag the matching resources with `Project=Atlas`, then write **one** policy that allows access only when the principal's `Project` tag equals the resource's `Project` tag.

The payoff shows up when a new project starts: **you write no new policies.** You tag the new identities and the new resources, and access flows automatically. ABAC is the scaling pattern for when the number of distinct permission sets would otherwise explode into hundreds of near-identical policies — the opposite of spelling out every ARN by hand.
