## Reading a scenario question & distractor patterns

SAA questions are **paragraphs, not one-liners.** A repeatable reading pattern turns time into accuracy:

1. **Find the verb first.** *"You need to **ensure / provide / minimise / prevent** …"* — that's the goal. Skim the setup until you find it.
2. **Highlight the constraints.** *least operational overhead · most cost-effective · minimum downtime · no public IP · across two Regions* — each constraint eliminates options.
3. **Match the constraint to a service, not a feature.** *"Global, edge-cached, with WAF"* → CloudFront. *"Reliable, in-order processing"* → SQS FIFO, not Standard or SNS.
4. **Eliminate by capability mismatch.** If an option simply *can't* do what's asked (Classic LB has no host-based routing; DynamoDB has no SQL JOINs), strike it — usually two of four fall away immediately.

### Distractor patterns

- **"Minimum operational overhead"** → the **managed** option (Lambda > EC2, Fargate > self-managed ECS, Aurora Serverless > Postgres on EC2).
- **"Encrypt at rest with a customer-managed key"** → **KMS CMK**, not the AWS-managed default.
- **"Most cost-effective" + "infrequent access"** → S3 Standard-IA or Glacier — check *first-byte latency* tolerance to choose.
- **CLB / NAT instance / EC2-Classic** → usually wrong in modern questions; ALB / NAT Gateway / VPC are the modern primitives.
- **Multi-AZ vs. Read Replica** → Multi-AZ = **failover** (not read scaling); read replicas = **read scaling** (not automatic failover). *"Increase read throughput"* wants **read replicas**.

The meta-skill: **the exam rewards elimination over recall.** You rarely need the single perfect fact — you need to strike the two options that violate a stated constraint, then choose between the survivors on the explicit preference.
