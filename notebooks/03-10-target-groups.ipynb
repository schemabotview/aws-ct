## Target groups, health checks & draining

A **target group** is the unit of routing behind a load balancer — it holds the targets and their health config, and it can be reused across listeners and even multiple load balancers. Each group has a **health check**: a protocol and path (e.g. HTTP `/health`) probed on an **interval** (default 30 s), with a **healthy threshold** (consecutive passes to bring a target into service — 5 for ALB) and an **unhealthy threshold** (failures to pull it out — 2 for ALB). Only healthy targets receive traffic.

When a target is removed — a deploy, a scale-in, a failure — the load balancer doesn't cut it off mid-request. **Deregistration delay** (connection draining) stops sending *new* requests but waits up to a set time (default **300 s**) for in-flight requests to finish. Lower it (say 30 s) for stateless services; keep it high for long-lived or WebSocket connections.

One more setting the exam tests: **cross-zone load balancing**, which spreads traffic evenly across targets in *all* AZs rather than only within the LB node's own zone. It's **on by default and free on ALB**, but **off by default on NLB and GWLB**, where enabling it incurs cross-AZ data-transfer charges.
