## AWS Lambda — the execution model

**Lambda** runs your code in response to events, scales automatically from zero to thousands of concurrent executions, and bills **per millisecond**. You provide a function; AWS provides the runtime, the network, the scaling, and integration with dozens of event sources. A function is a **handler** (the entry point AWS calls per invocation), a **runtime** (Python, Node.js, Java, Go, .NET, or custom), an **execution role** (its IAM permissions), and **configuration** (memory, timeout, env vars, network).

When an event arrives, Lambda finds or creates an **execution environment** — a sandboxed **Firecracker micro-VM** running your runtime and code. The lifecycle: `init code → handler(event, context) → response`. **Init code runs once** per environment — use it to set up DB connections, SDK clients, cached config, loaded models — and is **reused** across later invocations on the same warm environment. The **handler runs once per invocation**, receiving the event and a `context` (request id, remaining time, function name).

That execution environment is the key unit: it's the unit of **caching**, of **cold-start cost**, and of **concurrency**. Every Lambda concept that follows — packaging, cold starts, concurrency, VPC — ties back to how these environments are created, reused, and torn down.
