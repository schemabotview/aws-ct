## Cognito — user pools & identity pools

Everything so far has been machine and traffic security. Human users also need to sign in. **Cognito** is two distinct services under one name.

### User Pools — *who are you?*

A managed **user directory** with sign-up and sign-in. It handles:

- Username/email + password with configurable password policies
- **Social identity providers** — Google, Facebook, Apple, Amazon — via federation
- **Enterprise federation** — SAML 2.0 and OIDC (Okta, Azure AD, ADFS, Auth0)
- **MFA** (SMS, TOTP), email/phone verification, and **adaptive authentication** (risk-based MFA on suspicious sign-ins)
- **Hosted UI** — a pre-built, brandable sign-in page so apps ship no login form

On successful sign-in the User Pool returns three **JWT tokens** — an ID token (who the user is), an access token (what they can call), and a refresh token (to get new ones).

### Identity Pools — *what AWS resources can you reach?*

An Identity Pool takes an identity token (from a User Pool, or directly from Google / Facebook / SAML / OIDC) and exchanges it via STS `AssumeRoleWithWebIdentity` for **temporary AWS credentials**. The client then calls AWS directly — upload to S3, query DynamoDB — with **no backend proxy**.

- Each pool wires to **two IAM roles** — an **authenticated** role and an **unauthenticated (guest)** role for anonymous users.
- Supports **role-based access control**, mapping users to different IAM roles by their attributes.

The pairing: **User Pool authenticates → Identity Pool authorises to AWS.** Sign-in and a directory → User Pool. Temporary AWS credentials for a client app → Identity Pool.
