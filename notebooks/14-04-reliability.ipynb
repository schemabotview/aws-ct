## Reliability

The pillar about *meeting your uptime and recovery targets*. The checklist, pulling modules 03, 08, and 13 together:

- **Distributed across AZs** for every production tier — ASGs span all AZs, RDS Multi-AZ, ElastiCache Multi-AZ, ALB cross-zone load balancing.
- **Stateless compute** — sessions in ElastiCache, files in S3, secrets in Secrets Manager. Local disk holds nothing of value, so any instance can be replaced.
- **Automatic recovery** — Auto Scaling self-heals failed instances; ALB health checks remove bad targets; EC2 Auto Recovery handles host failures; the ECS scheduler replaces failed tasks.
- **Service quotas managed** — request increases *before* you hit them; alarm on quota usage via the Service Quotas console.
- **DR strategy chosen and tested** — Backup & Restore, Pilot Light, Warm Standby, or Active-Active per workload, with **quarterly drills** (module 13).
- **Backpressure & retries** — SQS to absorb spikes, exponential backoff with jitter, idempotency so a retry is safe.

The design principles: **automatically recover from failure, test recovery procedures, scale horizontally, stop guessing capacity, manage change through automation.**

The mental model: **assume everything fails, and make failure a non-event.** A single instance, AZ, or dependency dying should degrade gracefully, not page anyone. The cue: *"survive an AZ outage,"* *"no single point of failure,"* *"recover automatically,"* *"meet an RTO/RPO"* → this pillar.
