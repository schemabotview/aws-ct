## WAF & Firewall Manager

### WAF — Web Application Firewall

WAF is a **Layer 7** filter for HTTP/HTTPS traffic. It inspects requests against rules you define and **allows, blocks, counts, or CAPTCHA-challenges** them before they reach your application. It attaches to **CloudFront** (global, at the edge), **Application Load Balancer**, **API Gateway**, **AppSync**, and **Cognito User Pools**.

The top-level resource is a **Web ACL** — a container of rules with a default action (Allow or Block) for unmatched requests. Rules run in priority order; the first match wins.

| Rule type | Matches |
|---|---|
| **IP set** | specific IPs / CIDR ranges |
| **Geo match** | by country |
| **Rate-based** | IPs exceeding a per-5-min threshold |
| **String / regex** | URI, query string, headers, body |
| **SQL injection / XSS** | attack patterns in named fields |

**Managed Rule Groups** are pre-built, AWS- or vendor-maintained rule sets — the **Core Rule Set**, **Known Bad Inputs**, **Bot Control** — so you get broad coverage without hand-writing rules.

### Firewall Manager

Firewall Manager centralises security policy across an **AWS Organisation** — WAF Web ACLs, Shield Advanced subscriptions, security-group baselines, Network Firewall, and Route 53 DNS Firewall. Define a policy once in the management (or delegated-admin) account and it applies and enforces across all member accounts.

Two patterns: **enforce a baseline** (every internet-facing ALB must have the Core Rule Set attached) and **auto-onboard** (a new account inherits every policy on day one). The cue: *"apply WAF/Shield across all accounts,"* *"new accounts must inherit automatically"* → Firewall Manager.
