## Detective services — GuardDuty, Macie, Inspector, Detective

Preventive controls block known threats; you can't anticipate everything. **Detective controls** watch for unusual behaviour after the fact — none of them block traffic; all surface findings for you (or Security Hub) to act on.

- **GuardDuty** — continuous **threat detection** across CloudTrail, VPC Flow Logs, and DNS query logs (optionally S3 data events, EKS audit logs, EBS malware scanning). ML models flag anomalies — unusual API calls, known-bad-IP traffic, crypto-mining, credential exfiltration. Enable once at the Organisation level for every account; findings come out as **EventBridge events**.
- **Macie** — **sensitive-data discovery in S3**. Scans buckets for PII, financial data, credentials, and health data using regex + ML, producing a per-bucket inventory and findings when data is exposed. The answer for compliance audits over S3 data lakes.
- **Inspector** — continuous **vulnerability assessment** for **EC2, ECR container images, and Lambda**. Scans for known CVEs in OS packages and libraries and for unintended network exposure; runs automatically on resources it discovers.
- **Detective** — **incident investigation**. Pulls in CloudTrail, VPC Flow Logs, and GuardDuty findings and stitches them into a graph of principals, resources, IPs, and time. After a GuardDuty finding, Detective answers "what else did that principal touch, when, from where?"
- **Security Hub** — sits over all four, aggregating findings (plus IAM Access Analyzer, Config, third-party) into one place with CIS / PCI / FSBP **compliance views**.

Cue map: *unusual API/known-bad IP* → GuardDuty · *sensitive data in S3* → Macie · *vuln-scan EC2/ECR/Lambda* → Inspector · *investigate a finding* → Detective · *aggregate + compliance* → Security Hub.
