## RDS overview

**RDS** runs the database engine on EC2 behind the scenes, with EBS for storage, exposing a **connection endpoint**, a **maintenance window**, and a small set of config knobs. Everything else — patching the OS, upgrading the engine, replacing failed hardware, taking backups, setting up replicas, flipping Multi-AZ on failure — is AWS's job. The trade: **no SSH, no OS access**, so any feature AWS hasn't certified (a custom extension, a kernel monitoring agent, Oracle RAC) is out of reach.

RDS supports **six engines**: **MySQL, PostgreSQL, MariaDB, Oracle** (BYOL or License Included), **Microsoft SQL Server**, and **IBM Db2**. All six get the same managed plane — automated backups, Multi-AZ, read replicas, Performance Insights, parameter and option groups — though the exact feature matrix varies per engine.

The mental split is simple: **RDS owns the infrastructure and operations; you own the schema, the queries, and the data**. That boundary is the whole value proposition — and its one limit. If your workload needs OS-level control or an uncertified feature, you're back on EC2; otherwise RDS is the default relational tier, and the rest of this act is the operational features layered on top (backups, availability, security, connection pooling).
