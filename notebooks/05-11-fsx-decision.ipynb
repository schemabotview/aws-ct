## FSx, Storage Gateway & the storage decision

**FSx** is a family of managed third-party filesystems, each for a workload EFS doesn't fit. **FSx for Windows File Server** speaks **SMB**, stores in **NTFS**, integrates with **Active Directory** — the answer whenever Windows/SMB/AD enters the question (multi-AZ runs active + standby; daily backups to S3). **FSx for Lustre** delivers **hundreds of GB/s** and sub-ms latency for **HPC and ML** — and links to an **S3 bucket** so files appear on first access and results export back. **FSx for NetApp ONTAP** (multi-protocol NFS+SMB+iSCSI) and **FSx for OpenZFS** are lift-and-shift answers when you already run NetApp or ZFS.

**Storage Gateway** bridges on-prem to AWS, presenting cloud storage as local. **S3 File Gateway** (NFS/SMB → S3) lands on-prem files directly in S3; **FSx File Gateway** caches a central FSx for Windows at a branch; **Volume Gateway** (iSCSI, Cached or Stored) gives block storage backed by S3/EBS snapshots; **Tape Gateway** (iSCSI VTL → S3/Glacier) replaces physical tape libraries with backup software unchanged.

The file-storage decision, since names don't reveal fit: **Linux NFS, shared → EFS**; **Windows SMB/AD → FSx for Windows**; **HPC/ML throughput → FSx for Lustre**; **NetApp/ZFS lift-and-shift → ONTAP/OpenZFS**; **on-prem files into S3 → S3 File Gateway**. The default answer to most generic "shared file storage" questions: **EFS for Linux, FSx for Windows.**
