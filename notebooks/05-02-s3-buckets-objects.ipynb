## S3 — buckets & objects

S3 stores **objects** in **buckets**. A bucket is a container with a **globally unique** name (3–63 chars, lowercase, no underscores); an object is a file plus metadata, addressed by a **key** that is just a string. The key `images/2024/photo.jpg` looks like a path, but **S3 is flat** — the slashes are part of the name, and the console only *renders* prefixes as folders; there are no real directories underneath.

The numbers that matter: a bucket lives in **one region** (data never leaves unless you replicate); objects run **0 bytes to 5 TB**; a single PUT tops out at **5 GB**, so use **multipart upload** above that (recommended above 100 MB). Metadata is system headers (Content-Type, ETag) plus up to 2 KB of user-defined headers. There's **no limit** on object count, and request capacity scales **per prefix**.

That minimal core — buckets, objects, string keys — is the whole mental model. Every other S3 feature (storage classes, security, lifecycle, replication, access points) bolts onto it. The flatness is exactly what lets S3 scale to any size cheaply; the rest of this act walks those features in the order you meet them running a real bucket.
