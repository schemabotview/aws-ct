## Policies — shape, types & the five policy slots

Every IAM policy is a JSON document, and every statement boils down to four fields: **Effect** (`Allow` or `Deny`), **Action** (the API operation, like `s3:GetObject`), **Resource** (the ARN it applies to), and an optional **Condition** (context checks — IP, MFA, region, tag).

Policies come in three managed flavors: **AWS managed** (prebuilt, not editable — `ReadOnlyAccess`), **customer managed** (your own, reusable and version-controlled — the default best practice), and **inline** (one-off, bound to a single identity, deleted with it).

The subtle part is *where* a policy attaches — there are **five slots**, ordered here by blast radius, broadest first:

1. **Service Control Policy (SCP)** — on an Org root, OU, or account. An org-wide *ceiling*; caps every principal in scope, **including member-account root**.
2. **Permission boundary** — on a user or role. A per-identity *ceiling*; caps what identity policies can ever grant.
3. **Identity-based** — on a user, group, or role. The job description — **where most `Allow`s come from.**
4. **Resource-based** — on the bucket, queue, key, or function. A lock on the asset itself; uniquely includes a `Principal`, enabling cross-account access *without* role assumption.
5. **Session policy** — passed at `AssumeRole` time. A per-session filter that shrinks a role for one session.

Slots 1, 2, and 5 only *cap*; slots 3 and 4 *grant*. The effective permission is the **intersection**: every cap must allow it, at least one grant must allow it — and an explicit `Deny` anywhere kills it.
