## Operational Excellence

The pillar about *running the system well day to day* — can you ship changes safely, respond to incidents, and improve from what you learn? The concrete checklist that emerges from the previous modules:

- **Infrastructure as code** for everything — CloudFormation, CDK, or Terraform. Manual console changes are flagged (AWS Config) and reverted.
- **Pipelines** — CodePipeline / CodeBuild / CodeDeploy or GitHub Actions, authenticating via **OIDC federation** (no long-lived IAM keys in CI).
- **Safe deployment** — **blue/green** for ECS and Lambda (CodeDeploy traffic shifting), **canary** for ALB, with **automated rollback on alarm**.
- **Observability by default** — CloudWatch Logs / Metrics / Alarms wired from the start; CloudTrail enabled org-wide; AWS Config recording.
- **Runbooks** — Systems Manager Automation documents for common operations, so responses are repeatable, not tribal.

The design principles behind the checklist: **perform operations as code, make frequent small reversible changes, refine procedures often, anticipate failure, and learn from every operational event.**

The cue in a scenario: *"reduce manual effort,"* *"deploy without downtime,"* *"roll back automatically,"* *"repeatable operations"* → this pillar, answered with IaC + pipelines + CloudWatch-driven automation. It leans on the observability stack from module 12 and the deployment safety of blue/green traffic shifting.
