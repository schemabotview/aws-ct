## AWS global infrastructure — the hierarchy

AWS's footprint is a set of **nested layers**, and every layer trades **breadth for proximity to the user**. The whole map falls out of three pressures the cloud handles at once: **latency** (users are worldwide, the speed of light is fixed), **sovereignty** (data has legal homes), and **failure isolation** (one building losing power must not take your service down).

From broadest to closest:

- **Region** — an independent geographic area with its own cluster of data centers. 30-plus worldwide; runs all AWS services.
- **Availability Zone (AZ)** — one or more discrete data centers inside a region. 3–6 per region; runs all regional services.
- **Edge location / PoP** — a small facility that terminates user connections close to the user. 600-plus; runs CloudFront, Route 53, WAF, Shield, Lambda@Edge.

Three more variants push AWS even closer when a region is too far: **Local Zones** (into a metro city), **Wavelength** (into a 5G carrier network), and **Outposts** (into *your* data center). The next sections zoom into each layer.
