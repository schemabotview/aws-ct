## SQS vs. SNS vs. EventBridge — the messaging triangle

Three services that all decouple producers from consumers and look alike — picking is a frequent decision. **SQS** is a **queue (pull)**: one consumer per message, **persistent** up to 14 days, no filtering (DLQ only), very cheap, very high throughput, sources are your applications. **SNS** is **pub/sub (push)**: many consumers at once, **no persistence**, per-subscription attribute filters, cheap, very high throughput, sources are your applications. **EventBridge** is an **event bus with rules**: many targets via **rich pattern matching**, optional archive, sources include **AWS services + SaaS + custom** — a little more cost and rule overhead.

The decision tree: **decouple from a single consumer + buffer traffic → SQS**; **one event → many subscribers, simple delivery → SNS** (often paired with SQS for durable fan-out); **route AWS/SaaS events with complex matching, or many-to-many across accounts/regions → EventBridge**.

The mental model: **SQS buffers work for one worker; SNS broadcasts to many now; EventBridge routes by content from many sources**. SQS+SNS is the durable fan-out workhorse; EventBridge is the richer router when the *source* is AWS/SaaS or the *matching* is complex.
