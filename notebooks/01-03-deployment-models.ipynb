## Deployment models — public, private, hybrid, multi-cloud

Where the infrastructure actually lives — a separate question from the service model.

**Public cloud** is provider-owned and multi-tenant: you share physical hardware with other customers, isolated logically. This is the default for everything in this course.

**Private cloud** is single-organization infrastructure — dedicated hardware, no other tenants. You see it where regulation or policy demands it: banks, government, sensitive workloads. On AWS this shows up as **Outposts** and **Dedicated Hosts**.

**Hybrid** connects a public cloud to private or on-premises infrastructure over a dedicated link — a VPN or Direct Connect. Most large enterprises live here during and after migration: some systems stay on-prem, the rest run in AWS.

**Multi-cloud** spreads workloads across more than one public provider — AWS plus Azure or GCP. The usual reasons: avoiding vendor lock-in, or using a specific service only one provider offers.

For the exam and for most real work, **public is the default**; the others are each an answer to a specific constraint — residency, latency to on-prem systems, or lock-in avoidance.
