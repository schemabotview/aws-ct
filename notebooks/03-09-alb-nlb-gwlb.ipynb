## ALB, NLB & GWLB — picking the load balancer

The **ALB** terminates the HTTP connection, parses the request, and routes on its content. Its object model is a **listener** (port + protocol) holding **rules** evaluated in priority order; each rule's **conditions** (path, host, header, query, source IP, method) pick an **action** (forward, weighted forward, redirect, fixed response, authenticate) pointing at a **target group**. Targets can be `instance`, `ip` (on-prem or peered), or `lambda`. It's the default for web apps, APIs, and microservices.

The **NLB** operates at Layer 4 — it moves packets rather than parsing them, which makes it fast and protocol-agnostic. It gives a **static IP per AZ** (or an Elastic IP), **preserves the client's real source IP** (an ALB rewrites it via `X-Forwarded-For`), handles millions of requests per second at single-digit-millisecond latency, and is the **only type that supports UDP**. Reach for NLB when you need a static IP for firewall whitelisting, when traffic isn't HTTP (gaming, IoT, custom protocols), or when raw latency matters more than Layer-7 features.

The **GWLB** is niche: it sits inline as a transparent bump-in-the-wire for a fleet of third-party appliances, using **GENEVE on port 6081** to encapsulate packets to a firewall or IDS and back — the workloads never know it's there. Recognise it when a design says "deploy third-party firewalls scalably and transparently."
