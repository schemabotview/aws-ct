## NAT Gateway — outbound for private subnets

Private subnets often need **outbound** internet — package mirrors, software updates, AWS endpoints not reached privately — but must not be **reachable inbound**. A **NAT Gateway** is the standard answer: a managed component that **sits in a public subnet**, holds an **Elastic IP**, and translates outbound traffic from private subnets so it appears to come from that EIP. It's **stateful** (outbound only; inbound responses allowed automatically), auto-scales to **100 Gbps**, and costs an **hourly rate + per-GB** data processing. It **can't have a security group** attached.

The **HA caveat** is the exam favourite: a NAT Gateway lives in **one AZ**. Put a single NAT Gateway in AZ-a, and if AZ-a fails, **every** private subnet routed through it loses internet — even subnets in healthy AZs. The correct pattern is **one NAT Gateway per AZ**, with each AZ's private route table pointing at its own — so an AZ failure is contained.

(The older **NAT Instance** — a NAT-configured EC2 you manage — is legacy: you patch it, it's a single point of failure, and its throughput is the instance's. Use a NAT Gateway.)
