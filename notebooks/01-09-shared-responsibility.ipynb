## The shared responsibility model

Security on AWS is **split** between provider and customer, and the slogan captures the line: AWS secures **the cloud**, you secure what you put **in** the cloud.

**AWS is always responsible for** the physical buildings, the hypervisors, the networking gear, the host operating system, and the managed services they operate. **You are always responsible for** your data, your IAM users, roles, and policies, your firewall rules, and your encryption choices.

Where exactly the line sits **depends on how managed the service is**:

- **EC2** — AWS owns the hardware and hypervisor; *you* patch the guest OS, run the app, and set network rules.
- **RDS** — AWS also patches the OS and database engine; *you* own users, queries, and backup configuration.
- **S3** — AWS guarantees the hardware, OS, and storage durability; *you* own the data, bucket policies, and access controls.

The line **moves up the stack as the service gets more managed** — but responsibility for *what you put in* and *who you let see it* never leaves you. Module 12 returns to this in depth.
