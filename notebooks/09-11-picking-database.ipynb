## Picking the right database

A **cue-to-service** cheat sheet spanning the relational chapter and this one — match the cue, get the service.

- **Key-value/document at any scale, single-digit ms, serverless** → **DynamoDB**
- **DynamoDB but microseconds, read-heavy** → **DAX**
- **MongoDB-compatible** → **DocumentDB** · **Cassandra-compatible** → **Keyspaces**
- **Graph** (social, fraud, recommendations) → **Neptune**
- **Time-series** (IoT, sensors, metrics) → **Timestream**
- **Redis as a durable primary** → **MemoryDB**
- **Immutable audit trail (centralized)** → **QLDB** · **decentralized ledger** → **Managed Blockchain**
- **OLAP, BI, TB–PB SQL analytics** → **Redshift** · **querying S3 beside cluster data** → **Spectrum**
- **Serverless SQL on S3, no warehouse** → **Athena** · **serverless Spark ETL + catalog** → **Glue**
- **Managed Hadoop (Spark/Hive/Presto/HBase)** → **EMR** · **full-text search + log analytics** → **OpenSearch**
- *(relational OLTP → RDS / Aurora, from module 08)*

The meta-pattern: each service answers **one** data shape or access pattern. **Real architectures stack two or three** — DynamoDB + OpenSearch (operational + search), Redshift + Athena + Glue (warehouse + lake), DynamoDB + DAX (hot reads). The question is rarely "which database" but **"which mix."**
