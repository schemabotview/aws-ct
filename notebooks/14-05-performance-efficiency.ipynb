## Performance Efficiency

The pillar about *delivering on performance targets without overspending*. The checklist:

- **Pick the right compute** — **Lambda** for sporadic event work, **Fargate** for steady containers, **EC2** for sustained or custom workloads, **Graviton** (`m6g`, `c7g`) for ~20% better price/performance on most x86 workloads.
- **Pick the right data engine per access pattern** — RDS/Aurora for OLTP, DynamoDB for key-value at any scale, Redshift for warehouse, S3 + Athena for ad-hoc lake, OpenSearch for search, ElastiCache for hot data.
- **Cache aggressively** — **CloudFront** at the edge, **ElastiCache** near the application, **DAX** in front of DynamoDB when latency demands.
- **Asynchronous patterns** — SQS, SNS, EventBridge, Step Functions decouple latency-sensitive paths from slow downstream work.
- **Auto-scaling with cooldowns** — scale on the right metric (CPU, queue depth, request count), tuned so it reacts without thrashing.

The design principles: **democratise advanced tech (use managed services), go global in minutes, use serverless, experiment often, and match the technology to the workload with mechanical sympathy.**

The core move is **match the tool to the access pattern** — the single most-tested idea on the exam. There is no universal "best" database or compute; there is only the best fit for *this* read/write shape, latency target, and scale. The cue: *"lowest latency,"* *"handle unpredictable spikes,"* *"the right database for this access pattern"* → this pillar.
