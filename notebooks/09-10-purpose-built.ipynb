## Purpose-built databases — the rest of the zoo

A handful of databases exist because **both relational and DynamoDB are wrong shapes** for specific workloads. Each is a one-liner from being unfit for everything else — the value is **cue → service**.

- **DocumentDB** — MongoDB-compatible **document** store (Aurora-style storage: six copies across three AZs, auto-grow to 128 TB, up to 15 replicas). Cue: *MongoDB, document, variable-structure catalog*.
- **Neptune** — managed **graph** DB (nodes + edges; Gremlin, openCypher, SPARQL). Cue: *graph, social network, fraud/link analysis, recommendations, friend-of-friend, knowledge graph*.
- **Keyspaces** — serverless **Cassandra**-compatible. Cue: *Cassandra, CQL*.
- **Timestream** — purpose-built **time-series**. Cue: *IoT, sensors, metrics over time*.
- **MemoryDB** — **Redis as a durable primary** store (not just a cache — multi-AZ durable). Cue: *Redis as the database of record*.
- **QLDB** — immutable, cryptographically **verifiable ledger** (centralized audit trail). Cue: *tamper-proof audit log, verifiable history*. (For a **decentralized, multi-party** ledger → **Managed Blockchain**.)

The pattern is **cue-to-service, not service-to-capability**: you don't learn every feature — you learn the one phrase in a scenario that jumps you to the right service.
