## Transit Gateway

When the number of VPCs that need to talk grows past a handful, peering's full mesh stops being sustainable. **Transit Gateway (TGW)** replaces the mesh with a **hub-and-spoke**: every VPC attaches **once** to the TGW and the TGW routes between them. Ten VPCs that would need **45** peering connections need **10** attachments.

TGW is **regional** but takes rich attachments: VPCs, **Site-to-Site VPN**, **Direct Connect Gateways**, and **other TGWs** for inter-region peering. Routing is **transitive by default** — any attachment can reach any other — which raises the obvious question of how you keep environments apart.

The answer is **TGW route tables**: a TGW has multiple route tables, and each attachment associates with one. A **dev** attachment on the dev route table sees other dev VPCs but not prod; the **prod** route table sees prod VPCs and the on-prem attachment but not dev. That's how you build one **global network with enforced segmentation** — the reason TGW is the backbone of almost every multi-VPC/hybrid design at scale.
