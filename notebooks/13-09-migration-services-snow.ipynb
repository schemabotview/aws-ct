## Migration services & the Snow family

The R tells you the strategy; these services move the bytes.

- **Application Migration Service (MGN)** — the modern **lift-and-shift** tool (replaces the deprecated Server Migration Service). Install a lightweight agent on each source server (physical, VMware, Hyper-V, cross-cloud); MGN **continuously block-level replicates** disks to a low-cost EBS staging area. When ready, launch test instances or cut over — the source becomes a **byte-identical EC2 instance in minutes**.
- **Database Migration Service (DMS)** — moves the database tier. **Homogeneous** (Oracle → RDS Oracle) keeps the schema; **heterogeneous** (Oracle → Aurora PostgreSQL) uses the **Schema Conversion Tool (SCT)** to translate schema + stored procedures, then DMS streams data with ongoing **CDC** so source and target stay in sync until cutover. **Near-zero-downtime.**
- **DataSync** — network-based **bulk file transfer** (on-prem NFS/SMB/HDFS/object → S3, EFS, FSx), with a managed agent, scheduling, and integrity checks.

When the network won't do, AWS ships you hardware — the **Snow family**:

- **Snowcone** — small, rugged, **8 TB**, battery-powered; IoT / field collection + some compute.
- **Snowball Edge** — the workhorse: **Storage Optimized (~80 TB)** or **Compute Optimized** (more vCPU/GPU for edge compute).
- **Snowmobile** — a 45-ft shipping container, **100 PB** per unit, for exabyte-scale data-centre evacuations.

Rule of thumb: **if shipping a Snowball beats your network link over the migration window, ship it.** (500 TB on a 100 Mbps link ≈ 460 days online.) Encryption is end-to-end with **KMS keys you control**.
