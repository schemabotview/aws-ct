## S3 storage classes

Every S3 class shares the same **eleven-nines durability** — they differ on **availability, retrieval cost, and minimum storage duration**. **Standard** is the hot default (instant, no minimum). **Intelligent-Tiering** auto-moves objects between tiers for **unknown or shifting** access patterns — the safe choice when you don't know the pattern. **Standard-IA** (30-day minimum, retrieval fee) fits backups and monthly-accessed data; **One Zone-IA** is the same but **single-AZ** — cheaper, only for **reproducible** data, because losing that AZ loses the data.

The **Glacier** tiers are for archives, trading retrieval time for cost: **Glacier Instant Retrieval** (90-day min, milliseconds — archives read once a quarter), **Glacier Flexible Retrieval** (minutes to hours — backups/DR), and **Glacier Deep Archive** (180-day min, 12–48 hours — long-term compliance).

Two traps: **One Zone-IA** only for data you can regenerate (single AZ = no redundancy), and each IA/Glacier class has a **minimum billed duration** — delete early and you still pay the full 30/90/180 days. Match the class to how hot the data is *and* how long it'll live; lifecycle rules (next sections) automate the transitions.
