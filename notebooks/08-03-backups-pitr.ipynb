## RDS backups & point-in-time restore

RDS runs **two backup mechanisms** at once. **Automated backups** are on by default: a daily snapshot in the maintenance window plus **transaction logs shipped to S3 every 5 minutes**, retained **1–35 days** (0 disables). Within that window you get **point-in-time restore (PITR)** to any second — the snapshot is the base and the logs replay forward. Automated backups are **tied to the instance** — delete the instance and they go too, unless you take a final snapshot.

**Manual snapshots** are triggered by you or automation (Lambda, EventBridge, AWS Backup) and **persist until you delete them** (instance deletion doesn't touch them). They can be **copied cross-region** (the standard DR pattern) and **shared with other accounts** (optionally re-encrypted).

The behaviour to remember: **restoring always creates a new instance** — PITR doesn't rewind the database in place, it spins up a parallel one you then repoint the app at. (Contrast this with **Aurora Backtrack** later, which *does* rewind in place.) The two mechanisms cover different needs: **automated + PITR** for "undo the last few hours," **manual snapshots** for long-term retention, cross-region DR, and account sharing.
