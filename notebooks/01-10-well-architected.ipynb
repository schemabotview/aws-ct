## The Well-Architected Framework — six pillars

The Well-Architected Framework is the lens AWS publishes for reviewing any workload. It isn't a checklist — it's a **vocabulary for trade-offs**, six pillars you weigh against each other:

- **Operational Excellence** — run and improve systems to deliver value: infrastructure-as-code, observability, runbooks, post-mortems.
- **Security** — protect data, systems, and assets: IAM, encryption, detective controls, incident response.
- **Reliability** — perform correctly and recover from failure: multi-AZ, Auto Scaling, DR planning.
- **Performance Efficiency** — use the right resources and keep them right as demand shifts: right-sizing, the right database for the workload.
- **Cost Optimization** — deliver value at the lowest price point: right-sizing, Reserved / Spot / Savings Plans, killing idle resources.
- **Sustainability** — minimize environmental impact: efficient regions, right-sizing, managed services.

You rarely maximize all six at once — tightening cost can trade against performance, hardening security can trade against operational speed. The value is having the **shared words** to name those trade-offs deliberately. Module 14 returns to each pillar with concrete service mappings.
