## Right-sizing and idle elimination

The first lever on every cost review: **resources not doing the work they're sized for**. Two patterns:

- **Over-sized** — an `m5.4xlarge` running at 8% CPU for 90 days. Drop to `m5.large` or a **Graviton** equivalent and save **60%+**.
- **Idle** — an EC2 instance getting no requests, forgotten after a project. A dev RDS running 24/7 but used only office hours. An EBS volume detached for months. An unused Elastic IP (which is billed when unattached).

Two AWS tools find these:

- **Compute Optimizer** — AWS's right-sizing engine. Analyses **14 days** of CloudWatch data for EC2, ASGs, EBS, Lambda, ECS, and RDS, and recommends a specific smaller (or differently-shaped) resource — *"switch to m6g.large, projected saving $X"* — accounting for performance impact.
- **Trusted Advisor** — flags idle resources in its **Cost Optimization** category: idle load balancers, underutilised EBS, low-utilisation EC2. (Full checks need Business/Enterprise Support.)

Beyond right-sizing, two structural levers:

- **Schedule non-prod** — stop dev/test instances nights and weekends (Instance Scheduler); ~70% of a 24/7 bill vanishes for office-hours workloads.
- **Storage lifecycle** — S3 **Intelligent-Tiering** and lifecycle rules move cold data to cheaper classes automatically; delete unattached EBS and old snapshots.

The mindset: **the cheapest resource is the one you turned off.**
