## S3 data protection — versioning, Object Lock, replication

**Versioning** preserves every PUT and DELETE. A new upload gets a new version ID; a delete just adds a **delete marker** as the latest version, with prior versions retrievable underneath — you permanently remove an object only by naming its exact version ID. Once enabled, versioning can only be **suspended, never disabled**, and versions accumulate storage cost — so a versioned bucket almost always wants a lifecycle rule to expire non-current versions. **MFA Delete** (root-only) requires an MFA token to permanently delete a version or change the versioning state. **Object Lock** is the strong guarantee: **WORM** (write-once-read-many) storage where objects can't be changed or deleted for a retention period — for compliance.

**Replication** asynchronously copies objects to another bucket — **CRR** (cross-region) for DR, latency, and residency; **SRR** (same-region) for log aggregation and cross-account copies. Both buckets need versioning, and a service role grants the read/write.

The surprises are what replication **doesn't** do by default: it **doesn't backfill** existing objects (use **S3 Batch Replication**), **doesn't replicate delete markers** (opt in — this stops accidental deletes cascading), **doesn't chain** (A→B doesn't trigger B→C), and skips Glacier objects. SSE-KMS objects need explicit key config on both sides.
