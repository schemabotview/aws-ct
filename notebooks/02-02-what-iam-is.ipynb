## What IAM is — the authorization engine

**IAM** — Identity and Access Management — is the service that decides, for **every single API call**, whether the caller is allowed to do what they're asking. A console click, a CLI command, a `boto3` call — they all funnel through the same authorization engine before anything happens.

Three properties define it:

- **Global** — no region selector. Users, groups, roles, and policies exist account-wide, not per region.
- **Default-deny** — if no policy *explicitly* allows an action, it is denied. Silence means no.
- **Free** — you never pay for IAM itself, only for the resources it grants access to.

IAM sits *underneath* every other AWS service. You don't "turn it on" — every request is already being evaluated by it. The rest of this module is really about the vocabulary IAM gives you to answer one question precisely: *who* may do *what*, on *which* resource, under *which* conditions.
