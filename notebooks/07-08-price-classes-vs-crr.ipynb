## Price classes & CloudFront vs. S3 Cross-Region Replication

CloudFront's per-GB cost varies by edge region — Europe and North America cheapest, Asia-Pacific and South America pricier. **Price classes** let you opt out of the pricier regions to save money, at the cost of users there being served from the next-nearest included edge (slightly higher latency). **All** = every edge worldwide; **200** = all except South America / Australia / NZ; **100** = North America + Europe only. Pick **All** for genuinely global low-latency; **100** for cost-sensitive apps whose users are mostly NA/EU. It **doesn't affect who can reach** the content — a South American user still loads a Class-100 distribution, just from a NA edge.

**CloudFront vs. S3 Cross-Region Replication** come up together because both cut latency for distant users, but they solve different problems. **CloudFront caches** objects at 600+ edges — read-only, TTL-based (stale until expired/invalidated), automatically global. **CRR copies** objects to another bucket in another region — near real-time, fully writable, only the regions you replicate to. **CloudFront** when the content is the same everywhere and you want it cached everywhere. **CRR** when you need the **data itself** living in another region (DR, regional reads, legal residency). They combine: replicate to a backup region and use a CloudFront **origin group** to fail over.
