## Availability Zones

An **Availability Zone**, or AZ, is one or more discrete data centers, each with its own power, cooling, and physical security. A region has **three to six** of them, named by suffix — `us-east-1a`, `us-east-1b`, `us-east-1c`.

Two properties make them useful:

- They sit **tens of miles apart** — far enough that a single fire or flood can't take more than one — yet
- they're joined by **high-bandwidth, low-latency fiber**, so the network between them is fast and effectively local.

That combination is the foundation of **high availability** on AWS. Running across two or three AZs inside a region means a single data center failure doesn't take your service down. Anywhere you meet **multi-AZ** in AWS — RDS Multi-AZ, an Auto Scaling group spread across zones, an ALB balancing across zones — this is the principle at work.
