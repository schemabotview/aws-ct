## Elastic Load Balancing — the four types

A single instance has two problems: it dies and the app is down until someone notices, and you can't ship a new version without downtime. The fix is to run **more than one** instance — which raises the question of how a user reaches *the app* rather than one instance by IP. That is a **load balancer**: it accepts connections at a **stable DNS endpoint** and spreads them across a fleet of healthy targets. Along the way it adds **health checks** (unhealthy targets stop getting traffic), **connection draining** (graceful removal mid-deploy), **TLS termination** (the cert lives on the LB, not every instance), and optional **stickiness**.

AWS offers four types. The **Application Load Balancer (ALB)** works at **Layer 7** — it understands HTTP/HTTPS and routes on content (path, host, header, query, method). The **Network Load Balancer (NLB)** works at **Layer 4** — it moves TCP/UDP packets without parsing them, giving a **static IP per AZ**, millions of requests per second, and a preserved source IP. The **Gateway Load Balancer (GWLB)** works at **Layer 3**, inserting third-party appliances (firewalls, IDS) transparently inline. The **Classic Load Balancer (CLB)** is legacy — only for old migrations.

The exam decision is almost always **ALB for HTTP, NLB for everything else** — the sections that follow drill into each.
