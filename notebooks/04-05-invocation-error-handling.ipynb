## Invocation models & error handling

Lambda is triggered three ways, and the model sets the retry and error rules. **Synchronous** — the caller waits for the response (API Gateway, ALB, Function URLs, direct SDK invokes); **no automatic retries** — the error goes back to the caller. **Asynchronous** — Lambda queues the event and returns immediately (S3, SNS, EventBridge); it **retries up to twice** with backoff, and persistent failures route to a Dead Letter Queue or a Destination. **Event source mapping (poll-based)** — Lambda polls the source and invokes with **batches** of records (SQS, Kinesis, DynamoDB Streams, Kafka/MSK); ordering and scaling depend on the source (SQS FIFO preserves order per message group).

Failure handling depends on the model. For **async**, a **Dead Letter Queue** (SQS/SNS) catches the failed payload after retries; **Lambda Destinations** are richer — they route both success and failure (with the full event + response) to SQS, SNS, EventBridge, or another Lambda. For **event source mappings**, retries run until the message/record expires, and **partial batch responses** (`ReportBatchItemFailures`) let the function report exactly which records failed so successful ones aren't reprocessed.

The exam signal: **sync = caller retries, async = 2 retries + DLQ/Destinations, poll = batch retries until expiry**.
