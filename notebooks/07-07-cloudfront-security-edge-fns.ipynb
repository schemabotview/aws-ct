## CloudFront security & edge functions

**Certificates**: CloudFront serves `*.cloudfront.net` with a default cert; for a custom domain you attach an **ACM certificate — which must live in `us-east-1`** (CloudFront reads certs from the global control plane there) regardless of origin/user location — a classic gotcha. **Viewer protocol policy** controls user→CloudFront (HTTP only / HTTPS only / redirect); **origin protocol policy** controls CloudFront→origin, independently — encrypt at least the viewer leg. **Geo restriction** allow/blocklists countries by IP geolocation (403 on block), *before* any cache lookup. **Signed URLs / signed cookies** grant time-bounded access to private content (paid downloads, premium streams).

CloudFront can run your code at the edge — two flavours. **CloudFront Functions**: constrained **JavaScript**, **viewer request/response only**, **sub-1 ms**, no network, ~6× cheaper, scales to **millions/sec** — for the trivially-cheap per-request stuff: URL rewrites, cache-key normalization, injecting security headers, simple JWT checks. **Lambda@Edge**: Node/Python, **all four trigger points** (viewer + origin), up to 5–30 s, **has network access**, up to 10 GB memory, scales to thousands/sec — when the logic needs a network call or a real runtime (auth against a service, complex origin rewrites).

Rule: **CloudFront Functions for cheap header/URL work at the viewer edge; Lambda@Edge when it needs the network or origin-side triggers.**
