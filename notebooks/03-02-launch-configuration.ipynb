## Launch configuration — what you choose

Every EC2 launch is the same set of fields, whether you click them in the console or pass them to an API. Eight matter:

- **AMI** — the OS image (Amazon Linux, Ubuntu, Windows, or a custom image).
- **Instance type** — the CPU + memory profile (e.g. `m7g.large`).
- **Key pair** — the SSH key for Linux, or the password-decryption key for Windows RDP.
- **Network** — which VPC and subnet, and whether to auto-assign a public IP.
- **Security group** — the instance-level firewall rules.
- **Storage** — the root EBS volume, plus any extra volumes.
- **IAM role** — permissions via an **instance profile**, so no long-lived keys ever sit on disk.
- **User data** — a shell script run **once**, as root, on first boot.

Two deserve emphasis. Attach an **IAM role** rather than baking access keys into the image — the instance metadata service vends rotating credentials automatically. And treat **user data** as your first-boot automation: install packages, pull code, register the instance. Everything after this section is really about choosing each of these fields *well*.
