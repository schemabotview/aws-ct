## Purchasing options — the EC2 cost lever

Purchasing options are the single most important EC2 cost lever, and each trades **commitment for discount**. **On-Demand** commits to nothing and pays full price — the right default for unpredictable or short-lived workloads. **Reserved Instances** commit to a specific instance shape for 1 or 3 years for roughly 40% up to ~72% off; an RI is a billing discount applied to matching on-demand usage, not a separate instance. **Savings Plans** are the modern replacement: you commit to a steady **spend rate** (say `$10/hour for a year`) rather than to specific instances, so the discount auto-applies as your fleet changes. **Spot** uses AWS's spare capacity at up to **90% off** — but AWS can reclaim the instance with a **2-minute warning**.

Two rules of thumb settle most exam questions. **Prefer Compute Savings Plans as the default commitment** — they flex across any region, any family, and even Fargate and Lambda; reach for Reserved Instances only when you need a capacity reservation or Marketplace resale. And **use Spot only for interruption-tolerant work** — batch jobs, CI runners, stateless web workers behind a load balancer, checkpointed ML training — never for databases or anything with un-checkpointable state.
