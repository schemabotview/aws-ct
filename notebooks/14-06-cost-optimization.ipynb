## Cost Optimization

Module 13 covered the cost levers in depth; the pillar's checklist, briefly:

- **Right-size weekly** — Compute Optimizer + Trusted Advisor + Cost Explorer.
- **Commit on the base** — Compute Savings Plans for 60–70% of the fleet; RIs for stable RDS / ElastiCache.
- **Spot for the burst layer** — batch, CI, stateless workers.
- **Storage lifecycle** — S3 Intelligent-Tiering by default; EBS gp3 over gp2; delete unattached volumes.
- **Idle elimination** — auto-shutdown for dev/test; scheduled scaling for off-hours.
- **Tag governance** — `cost-center`, `environment`, `owner`, `application` on everything; Cost Explorer pivots by tag.
- **Anomaly detection on day one** — catches the compromised-key cryptominer before the bill arrives.

The design principles: **adopt a consumption model, measure efficiency, stop spending on undifferentiated heavy lifting (use managed services), and attribute expenditure (tags + accounts).**

The insight that outlasts any single tactic: **cost optimisation is an operating discipline, not a one-time project.** A team that reviews spend weekly and adjusts beats a team that does one big cleanup a year and drifts. The cue: *"most cost-effective,"* *"minimise cost for a spiky workload,"* *"reduce spend without hurting performance"* → this pillar (and the level directly above the trade-off with reliability and performance).
