## Instance families, naming & T-series

Instance types group into **families** by workload shape — pick the family first, then the size. **General Purpose** (`t3`, `m7g`) balances CPU, memory, and network for web servers and small databases. **Compute Optimised** (`c7g`) gives the most CPU per dollar for batch, gaming, and HPC. **Memory Optimised** (`r7g`, `x2idn`) packs RAM per core for in-memory databases. **Storage Optimised** (`i4i`) delivers local NVMe throughput for NoSQL and warehouses. **Accelerated Computing** (`p4`, `g5`, `inf2`) adds GPUs and custom chips for ML.

The **name encodes everything**. In `m7g.xlarge`: `m` is the family, `7` the generation, `g` the processor (**Graviton**, AWS's ARM chip; `i` = Intel, `a` = AMD, plus `n` for network, `d` for local NVMe), and `xlarge` the size. Sizes roughly double vCPU and RAM at each step. **Graviton is usually ~20% cheaper and ~40% better performance-per-dollar** than x86 — the default first choice when your workload runs on ARM.

**T-series** (`t3`/`t4g`) is a special case: **burstable**. It runs at a baseline CPU percentage, earns credits when below baseline, and spends them to burst above it. **Unlimited mode** (the default) keeps bursting at extra cost; **Standard mode** caps at baseline. Perfect for spiky, mostly-idle workloads (low-traffic sites, dev boxes) — but wrong for sustained CPU, where you'll throttle or pay overage.
