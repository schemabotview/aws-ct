## Three storage shapes — object, block, file

Before you pick a storage service, pick the **shape of access** — there are three categories, and they aren't interchangeable. **Object storage** serves files over HTTP, scales to infinity, and has no filesystem semantics — that's **S3**. **Block storage** is a virtual disk attached to one machine for random block I/O — that's **EBS** (durable, network-attached) or **Instance Store** (local, ephemeral). **File storage** is a network filesystem many machines mount at once — that's **EFS** (Linux/NFS) or the **FSx** family (Windows, Lustre, and more).

The category is the decision that matters. Pick the wrong shape and nothing about the service fits your workload; pick the right one and the specifics are mostly tuning. Inside each category the service is largely obvious: S3 for object, EBS/Instance Store for block, EFS/FSx for file.

This module spends most of its time on **S3** — the centre of gravity of AWS storage, touched by almost every architecture as the durable home for backups, the staging area for analytics, the publisher of static sites, and the input/output of every data pipeline — then covers **EBS** and **Instance Store** for block, and **EFS/FSx** for file.
