## EBS snapshots, encryption & instance store

A **snapshot** is a point-in-time backup of a volume, stored in S3 behind the scenes (AWS-managed — you don't see the bucket). Snapshots are **incremental**: the first copies every used block, each later one copies only changed blocks. Restore creates a **new volume** in any AZ of the region — or another region after copying the snapshot across (the standard way to move a volume between AZs/regions). Three extensions: **Snapshot Archive** (~75% cheaper for rarely-restored compliance snapshots; 24–72 h to restore), **Recycle Bin** (holds deleted snapshots 1–365 days — the fat-finger safety net), and **Fast Snapshot Restore** (pre-warms a snapshot so new volumes have full performance immediately, no lazy-load latency).

**EBS encryption** is AES-256 backed by KMS, applied transparently — data at rest, snapshots, and volume↔instance traffic are all encrypted, with the Nitro hardware doing the work (negligible overhead). Snapshots of an encrypted volume are encrypted; volumes from an encrypted snapshot are encrypted. Flip **account-level default encryption** early — every new volume/snapshot is then encrypted automatically (existing ones aren't retroactively encrypted).

**Instance Store** is the opposite trade: **physically-attached NVMe** on the host — millions of IOPS, sub-ms latency — but **completely ephemeral**: stop, terminate, or lose the host and the data is gone (a reboot keeps it). It's included in the instance price, fixed-size, can't detach or resize. Use it for **caches, ML scratch, and replicated distributed systems** (Cassandra, Kafka) where node-local loss is normal.
