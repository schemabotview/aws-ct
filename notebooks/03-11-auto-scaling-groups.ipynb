## Auto Scaling Groups — launch templates & scaling

Behind a load balancer, the fleet size is still set by hand — traffic triples at lunch and the site slows, then goes quiet overnight while you pay for idle servers. An **Auto Scaling Group (ASG)** fixes that: it keeps a target number of instances running, replaces failed ones, and grows or shrinks the fleet on demand. Three numbers bound it — **minimum** (the floor), **desired** (what it actively maintains), and **maximum** (the ceiling) — and it spreads instances across the subnets you give it (one per AZ for HA), auto-registering each new instance with its **target groups** and deregistering terminated ones.

An ASG launches instances from a **launch template** — the full recipe: AMI, instance type(s), key pair, security groups, IAM role, user data, storage, IMDS options. Templates are **versioned** and support **mixed instance types** (On-Demand + Spot), which the legacy *launch configuration* can't — always use launch templates.

Scaling is driven by **policies**. **Target tracking** is the go-to: name a metric and a target (e.g. "average CPU at 50%") and the ASG adds or removes capacity to hold it. **Step scaling** changes capacity by different amounts at different alarm breaches; **simple scaling** is the older single-step form; and **scheduled scaling** changes capacity at known times (e.g. before business hours).
