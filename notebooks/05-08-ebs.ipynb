## EBS — block storage for EC2

An **EBS volume** is a **network-attached block device** an EC2 instance sees as a local disk. Because it's on the network, not inside the host, three properties follow: it can **detach from one instance and reattach to another** (in the same AZ), it **persists independently of the instance** (survives stop/terminate unless *Delete on Termination* is set — on by default for root, off for extra volumes), and it pays a tiny network-latency tax (negligible except for the most latency-sensitive work). EBS is **AZ-scoped** — a volume only attaches to instances in its AZ; crossing AZs or regions goes **through a snapshot**. Most volumes attach to one instance; **Multi-Attach** (io1/io2 only) allows up to 16.

**Volume types** split SSD vs HDD. **gp3** is the default for almost everything: 3,000 IOPS and 125 MB/s baseline regardless of size, with IOPS and throughput dialled **independently**, and ~20% cheaper than gp2 (which ties IOPS to size at 3 IOPS/GB). **io2 / io2 Block Express** are provisioned-IOPS SSD for I/O-intensive and the very largest databases (up to 256,000 IOPS). **st1** (throughput HDD) suits sequential big-data/logs; **sc1** (cold HDD) suits rarely-accessed bulk — both **not bootable**.

The short decision: **gp3 unless you have a reason** — io2 for demanding databases, st1/sc1 for sequential or cold bulk.
