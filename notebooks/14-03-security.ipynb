## Security

The pillar about *protecting data and identities against credible threats*. The non-negotiables, drawn from modules 02 and 11:

- **Identity** — **IAM roles, not users**, for everything; permission boundaries on developer roles; SSO via **IAM Identity Center**, not local IAM users.
- **Least privilege** — IAM Access Analyzer findings reviewed; policies sized to actual usage (CloudTrail-driven policy generation), not `*:*`.
- **Data encryption** — **KMS CMKs** for sensitive data; S3 default encryption; the EBS default-encryption flag on per Region; **TLS in transit everywhere**.
- **Secrets** — Secrets Manager with rotation, or SSM Parameter Store SecureString; **never** plain-text secrets in code, AMIs, or Lambda env vars.
- **Network controls** — private subnets for compute, public only for load balancers; security groups tight; NACLs as a coarse second layer.
- **Detection** — GuardDuty org-wide, Security Hub aggregating, Config rules for drift, CloudTrail immutable.

The design principles: **a strong identity foundation, traceability, security at all layers, automate security, protect data in transit and at rest, keep people away from data, prepare for events.**

Two ideas recur: **defence in depth** (identity + network + encryption + detection, so one failure isn't fatal) and **least privilege** (grant the minimum, expand only on evidence). The cue: *"encrypt at rest with your own key,"* *"restrict blast radius,"* *"detect unusual activity"* → this pillar.
