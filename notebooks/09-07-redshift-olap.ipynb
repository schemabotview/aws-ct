## Redshift — the OLAP warehouse

Transactional engines (RDS, Aurora, DynamoDB) do **many small queries against current state**. **Analytics** is a different shape — **few large queries**, full scans over billions of rows, aggregations and joins across years — and running those on a transactional engine grinds. The right tool is a **columnar, massively-parallel data warehouse**: on AWS, **Redshift**. Its job is **OLAP** (Online Analytical Processing) — monthly revenue rollups, cohort analysis, big multi-table joins — the opposite of the **OLTP** that RDS/Aurora serve.

The key word is **columnar**. A row store keeps each row's fields together (great for fetching one whole row by key); a columnar store keeps each **column's values together**, so `SELECT SUM(revenue) FROM orders` reads only the revenue column off disk instead of every column of every row — a huge win when a query touches a few columns out of fifty.

Architecturally a Redshift cluster is a **leader node** + **compute nodes**: the leader parses/plans the SQL and distributes work; each compute node runs its slice **in parallel** and ships results back — **MPP** (massively parallel processing), which is what lets it chew through TB–PB. Node families: **RA3** (managed storage — compute and storage scale independently, S3-backed with SSD cache — the default), and legacy **DC2** (local SSD, coupled). **Redshift Serverless** drops the cluster entirely — you pick a base capacity in **RPUs** and it scales automatically.
