## CloudWatch Agent & Insights variants

### The CloudWatch Agent

The **CloudWatch Agent** runs inside an EC2 instance (or on-prem host) and ships what the hypervisor can't see — **in-OS memory, swap, disk space, and filesystem-level disk I/O** — plus **log files** the OS or applications write. Config is JSON, often stored in **SSM Parameter Store** and pulled at startup; the instance's IAM role needs the `CloudWatchAgentServerPolicy` managed policy.

The agent is the answer to two classic scenarios:
- *"EC2 ran out of memory and nobody noticed"* → install the agent, push memory metrics, alarm on them.
- *"We need application logs in CloudWatch"* → point the agent at the log files; it streams them into log groups.

### Insights variants

- **Container Insights** — enhanced metrics + logs for **ECS** and **EKS**: per-container CPU, memory, network, disk, plus container stdout/stderr. Enable once per cluster; data lands in dedicated dashboards.
- **Lambda Insights** — function-level diagnostics: **cold-start counts, init duration, memory curves, CPU time, network I/O** per invocation — the deeper layer under standard Lambda metrics.

The through-line: standard metrics come from **outside** the instance; the agent and Insights variants get you **inside** — memory, containers, and function internals the hypervisor never sees.
