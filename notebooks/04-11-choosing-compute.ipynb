## Choosing a compute service

One higher-level option hides most of the choice: **App Runner**, the highest-level container path. Point it at an ECR image or a Git repo, and it builds, deploys, load-balances, terminates TLS, scales, and health-checks for you. Its sweet spot is a **small-to-medium web service that just needs to be online** behind HTTPS with autoscaling. You outgrow it when the architecture grows past one service or needs custom networking — then you drop down to ECS or EKS.

The full compute ladder, by workload shape:

- **Event-driven, short bursts (< 15 min)** → **Lambda**
- **HTTP API, simple routing** → **Lambda + API Gateway** or **App Runner**
- **Long-running service, no Kubernetes need** → **ECS on Fargate**
- **Long-running at high steady utilisation, cost-sensitive** → **ECS on EC2** + Savings Plans
- **Existing Kubernetes / multi-cloud / advanced scheduling** → **EKS**
- **Full OS control, special hardware** → **EC2 directly**

The progression Lambda → App Runner → ECS/Fargate → EKS → EC2 is a **ladder of increasing control and increasing operational responsibility**. Pick the **highest rung** that meets your requirements — every step down adds work you have to justify.
