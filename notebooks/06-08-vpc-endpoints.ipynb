## VPC endpoints & PrivateLink

A **VPC Endpoint** lets a VPC reach AWS services **privately** — no Internet Gateway, NAT, or VPN — which matters because NAT bandwidth and per-GB charges add up, and many regulated workloads can't let AWS-bound traffic leave the AWS network. Two types tell the whole story. A **Gateway Endpoint** serves **S3 and DynamoDB only**, is **free**, and works by adding a **route** to a prefix list in your route tables. An **Interface Endpoint** (powered by PrivateLink) serves **most AWS services**, works by creating an **ENI with a private IP** in your subnet, costs per hour + per GB, and optionally enables **private DNS** so the service's normal hostname resolves to the ENI. Rule: **Gateway Endpoint for S3/DynamoDB (free), Interface Endpoint for everything else.**

**PrivateLink** turns that same machinery around to expose **your own service** to other VPCs — no peering, no internet, no CIDR-overlap worry. The provider puts a **Network Load Balancer** in front and creates a **VPC Endpoint Service** registering it; consumers create an Interface Endpoint targeting that service, and traffic flows privately over the backbone from consumer IP to provider NLB. It's how **SaaS vendors** give thousands of customer VPCs a private connection, and how **internal microservices** are shared across accounts.
