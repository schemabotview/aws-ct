## The root user & account day one

When you create an AWS account, you get one identity for free: the **root user**. It owns the account and has unlimited, **un-restrictable** access — no policy, not even a Service Control Policy, can take permissions away from it. That power is exactly why day one is about *locking it away*, not using it.

A handful of tasks genuinely **require** root and can't be delegated: changing the account's email or root password, closing the account, restoring admin access if you've locked yourself out, registering as a seller in the Reserved Instance Marketplace, and a few billing settings. Everything else you do as an IAM identity instead.

**Root hygiene is non-negotiable:** give it a long, unique password; enable **hardware MFA**; **never issue access keys for root**; store the credentials somewhere physically safe; then leave it alone. Most healthy accounts go months or years without anyone signing in as root.

Everything that follows in this module — users, roles, policies, Organizations — exists so that nobody has to touch root to get real work done.
