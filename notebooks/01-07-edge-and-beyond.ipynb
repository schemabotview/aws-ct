## Edge locations & beyond-Region options

**Edge locations are not regions** — they don't run EC2 or RDS. They're smaller facilities in far more cities than regions, with a single job: terminate user connections close to the user. That's what powers **CloudFront** (cache and deliver content from the nearest edge), **Route 53** (answer DNS at the edge), **AWS WAF** (filter HTTP traffic), **AWS Shield** (DDoS protection), and **Lambda@Edge / CloudFront Functions** (run code to customize requests and responses).

When even a region is too far, three options push AWS services closer to a specific kind of user:

- **Local Zones** — AWS into a *metro city*, for live media production, interactive gaming, real-time finance.
- **Wavelength** — AWS into a *5G carrier network*, for connected vehicles, smart factories, mobile AR.
- **Outposts** — AWS into *your own data center*, for strict data residency or ultra-low latency to local industrial systems.

The common thread: each trades the full breadth of a region for **proximity** to somewhere a region can't reach.
