## EFS — shared file storage

**EFS** is a managed **NFS v4** filesystem that **many Linux instances mount at once** — across AZs, accounts, and even on-prem over Direct Connect/VPN. Unlike EBS, it's **regional**: data is replicated across AZs automatically, capacity grows and shrinks elastically, and you pay per GB with no provisioning. Clients connect through **mount targets** (one per AZ, each a private IP in your VPC), with security groups controlling access; every mount target resolves to the same filesystem, so a write through one is instantly visible through all. It's **Linux/POSIX-only** — Windows needs FSx for Windows. Typical uses: content shared across web servers, Linux home directories, shared config across an ASG, and container volumes for ECS/EKS (EFS CSI driver).

Two orthogonal knobs. **Performance mode** (set at creation): **General Purpose** (lowest latency — almost everything) or **Max I/O** (higher latency, scales to thousands of clients — big analytics/media). **Throughput mode** (changeable): **Elastic** (auto-scales, pay per GB transferred — the default for new filesystems), **Bursting** (baseline scales with size), or **Provisioned** (fixed, for a small dataset needing high throughput).

EFS has its own **tiering** by last-access time: **Standard → Standard-IA (~92% cheaper) → Archive (~95% cheaper)**, with optional promote-on-access. **EFS One Zone** is a single-AZ variant ~47% cheaper — for dev/test or already-backed-up data (an AZ loss means it's unavailable/lost).
