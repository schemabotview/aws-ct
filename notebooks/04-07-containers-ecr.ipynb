## Containers & ECR — the other compute model

Lambda gives you zero servers but constrains your runtime to AWS's choices. A **container** removes that limit: your application, its runtime, and its dependencies packaged into one **image that runs identically** wherever a container engine exists — booting in seconds, sharing the host kernel instead of virtualising hardware. Containers fit exactly where Lambda doesn't: **long-running** processes beyond the 15-minute cap, **sustained CPU/memory** where per-hour beats per-invocation, workloads needing **runtime control** Lambda doesn't expose, and **already-containerised** apps.

AWS runs containers three ways: **ECS** (AWS-native orchestration), **EKS** (managed Kubernetes), and **Fargate** as the serverless compute engine *underneath* either — the host is invisible.

Every container path starts at a **registry**. **ECR** (Elastic Container Registry) is AWS's managed image registry — the private, IAM-integrated equivalent of Docker Hub. One private registry per account per region, plus a public `public.ecr.aws`. It does **image scanning** (basic on push, or enhanced/continuous via Inspector), **lifecycle policies** to expire old versions, cross-region/account **replication**, and **KMS** encryption at rest. Auth is a **short-lived token** from `aws ecr get-login-password` — no long-lived registry credentials to manage.
