## CloudWatch vs. CloudTrail vs. Config — picking the right tool

The three overlap in casual talk but answer **distinct** questions. The map worth memorising:

| Question | Service |
|---|---|
| Is my CPU high? Is my Lambda erroring? Are requests slow? | **CloudWatch** — performance & health |
| Who deleted that S3 bucket? Who changed this IAM policy at 03:14? | **CloudTrail** — API audit log |
| What did my security group look like last Tuesday? Was it compliant? | **Config** — config history + rules |

### Composed patterns

| Cue | Answer |
|---|---|
| Alert when RDS CPU > 80% for 10 min | **CloudWatch alarm** on the RDS metric |
| Alert when someone calls `DeleteTrail` | **CloudTrail → CloudWatch metric filter → alarm** |
| Alert when an S3 bucket becomes public | **Config rule** `s3-bucket-public-read-prohibited` → SNS |
| Auto-fix a non-compliant security group | **Config rule + SSM Automation** remediation |
| In-OS EC2 memory / app logs | **CloudWatch Agent** |
| Watch the AWS bill | **billing alarm** on `EstimatedCharges` (us-east-1) |

The mnemonic that carries the whole module:

- **CloudWatch = how is it doing?** — metrics, logs, alarms, dashboards.
- **CloudTrail = who did what?** — every API call, retained and searchable.
- **Config = what changed, and is it compliant?** — config timeline + rules + remediation.

They **compose**: CloudTrail records an action, a metric filter turns it into a CloudWatch alarm; Config finds drift, SSM Automation fixes it.
