## CloudFront — caching close to users

**CloudFront** is AWS's CDN: cache content close to the user so most requests are answered from the **edge** rather than dragged to the origin. It runs from **600+ edge locations** in 100+ cities, with a tier of larger **regional edge caches** between edge and origin to absorb misses. A request flow: DNS resolves the distribution to the nearest edge → the edge checks its cache → **hit** returns immediately; **miss** fetches from the regional edge cache, then the origin over the AWS backbone, caching on the way back. Beyond caching, CloudFront **terminates TLS at the edge**, brings **free Shield Standard** DDoS protection and **WAF** integration, and speaks **HTTP/2 and HTTP/3** regardless of the origin.

An **origin** is where CloudFront fetches on a miss — an S3 bucket, S3 website endpoint, ALB, EC2, or any HTTP server. One distribution can have **multiple origins**, with cache behaviours routing path patterns to each.

For an **S3 origin**, the pattern is **Origin Access Control (OAC)**: keep the bucket **completely private** and let **only CloudFront** read it — CloudFront signs origin requests with SigV4, and the bucket policy allows the CloudFront service principal scoped to that distribution. No public access, no presigned-URL leakage. OAC **replaced OAI** (which lingers but lacks SSE-KMS support and newer regions) — default to OAC on every new distribution.
