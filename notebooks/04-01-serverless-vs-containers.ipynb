## Serverless, containers & EC2 — the trade-off

Module 03 gave you the canonical web tier: EC2 instances behind a load balancer, fronted by an Auto Scaling Group. It works — so why choose anything else? Three pressures push real workloads off pure EC2: **bursty demand** (an ASG ramps too slowly and you pay for the idle floor), **operational surface** (you don't want to patch OSes or bake AMIs for code that runs one second a minute), and **runtime packaging** (you want the exact same binary running on your laptop, in CI, and in prod).

Two models answer those. **Serverless** — you write code, AWS handles provisioning, patching, scaling, and HA; you pay only for compute actually used, and idle costs nothing. **Containers** — your code, runtime, and dependencies packaged into one portable image that runs anywhere a container engine does; AWS runs them native (ECS), Kubernetes-flavoured (EKS), or fully serverless (Fargate).

The trade-off is steady: **serverless wins on event-driven, intermittent, or unpredictable work** (automatic near-instant scaling, no idle cost, per-request pricing). **Long-running, stateful, or sustained-CPU work belongs on EC2 or containers** — at high steady utilisation, per-invocation pricing loses to per-hour. This module walks Lambda first, then containers (ECS, Fargate, EKS), and closes with the decision tree.
