## Streams, global tables, TTL & PITR

**DynamoDB Streams** is a **24-hour change log** of item-level modifications — each create/update/delete captured as an ordered (per PK) record, containing what you choose: keys only, new image, old image, or both. **Lambda consumes** streams via an event source mapping (auto-scaled to shards), which powers a lot of patterns: **event-driven downstream work** (fan a write to email/notifications), **search indexing** (replicate to OpenSearch), **audit logging** (append to S3 via Firehose), **materialized views** (a derived projection in another table), and **replication** — the basis of Global Tables.

**Global Tables** is DynamoDB's **multi-region, multi-active** replication — every region accepts reads *and* writes, changes replicate in ~1 s, conflicts resolved last-writer-wins. For globally distributed apps and multi-region DR.

Two operational features. **TTL** auto-deletes items after a per-item expiry (a Unix-epoch attribute) in the background (typically within 48 h) — **TTL deletes don't cost WCUs** and **appear in Streams** (so cleanup can hang off them); caveat: items may still show in reads until the sweep, so filter on the expiry attribute if strict. **PITR** is continuous backup — restore to **any second in the last 35 days** as a new table — the standard guard against corruption or a DELETE-without-WHERE; **on-demand backups** are the manual snapshot companion.
