## S3 lifecycle & cost management

**Lifecycle rules** move objects between storage classes and eventually delete them, automatically, by object age — scoped to the whole bucket, a prefix (`logs/`), tags, or a size range. Transitions only flow **downward** the cost ladder (Standard → IA → Glacier; you can't lifecycle an object back to a warmer class), with a **30-day minimum** before leaving Standard for an IA tier. The key expiration actions: **expire current versions** after N days (the retention cap), **expire non-current versions** (the lever that makes versioned buckets affordable), **delete expired delete markers** (housekeeping), and **abort incomplete multipart uploads** (a silent cost most accounts leak).

Three capabilities sit just outside the core and cut cost or effort. **S3 Select** runs **SQL against a single object** (CSV/JSON/Parquet) server-side, returning only matching rows — huge savings when you need a slice of a big file. **S3 Batch Operations** runs one action across **billions of objects** as a managed job (copy, invoke a Lambda per object, retag, restore, Batch Replication), driven by an S3 Inventory manifest. **S3 Storage Lens** is the org-wide analytics dashboard — 29+ metrics across every account and bucket, with cost recommendations (buckets missing lifecycle, incomplete uploads).

Together these are the cost-control toolkit: lifecycle to tier and expire, Select to read less, Storage Lens to find the waste.
