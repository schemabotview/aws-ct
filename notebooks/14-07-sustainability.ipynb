## Sustainability

The newest pillar — added in 2021 — is about *minimising the environmental impact of running cloud workloads*. The lens: **less compute consumed, fewer resources idle, more efficient hardware.**

Most of the checklist overlaps with Cost Optimization — efficiency saves both money and carbon:

- **Right-size and decommission idle** — the same lever that saves money saves carbon.
- **Modern instance generations** — Graviton processors consume meaningfully less power per unit of work than older x86.
- **Serverless and managed services** — share underlying hardware, raise utilisation, and cut per-customer footprint.
- **Storage tiering** — Glacier Deep Archive uses dramatically less energy than Standard for the same bytes at scale.
- **Region choice** — AWS publishes carbon intensity per Region; some run on significantly more renewable energy. For latency-flexible workloads, a greener Region is a free win.
- **Right-size the whole demand curve** — smaller data formats, efficient code, batching, and lifecycle deletion all reduce the work done.

The design principles: **understand your impact, set goals, maximise utilisation, anticipate and adopt more efficient hardware, use managed services, and reduce the downstream impact of your workloads.**

The one-line takeaway: **the greenest and cheapest architectures are usually the same architecture.** Sustainability rarely asks for a *different* design than cost — it asks you to count carbon alongside dollars. The cue: *"minimise environmental impact,"* *"reduce carbon footprint"* → this pillar (answered mostly with right-sizing, Graviton, serverless, and tiering).
