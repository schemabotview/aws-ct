## Aurora Serverless, Global Database & backtrack

**Aurora Serverless v2** auto-scales a cluster's **compute** in real time, in fine increments, by actual load. The unit is the **Aurora Capacity Unit (ACU)** (~2 GB RAM + proportional CPU/network), scaling in **0.5-ACU steps in seconds without dropping connections**; you set a **min/max ACU range**. Unlike the deprecated v1 (which paused/resumed by replacing the instance, with connection drops and a 10 s+ warmup), v2 scales the **live** instance. It fits **variable/unpredictable** load, dev/test, and multi-tenant SaaS, and can mix with provisioned instances in one cluster.

**Aurora Global Database** spans **multiple regions** as one logical cluster: one **primary region** for writes, up to **five secondaries** for reads, replicated at the **storage layer** with **sub-second lag across continents**. The DR headline: **RPO < 1 second, RTO < 1 minute** via managed failover (plus an unplanned manual detach if the primary region is unreachable). It's the strongest cross-region relational option and the pattern for global low-latency reads.

**Aurora Backtrack** rewinds an **Aurora MySQL** cluster to a past point **in place** — no snapshot restore, no new instance, no endpoint change — within a window up to **72 hours**, in seconds (the storage layer keeps a continuous change log). The use case: someone ran `DELETE FROM customers` without a `WHERE` — backtrack five minutes, done. **MySQL only.** (Aurora **Multi-Master** — multiple writers — exists but is niche and legacy; the standard single-writer + fast failover is right for almost everything.)
