## Identity federation

**Federation** lets external identities assume IAM roles **without ever becoming IAM users** — no per-person user, no long-lived password. Two protocols cover it: **SAML 2.0** for enterprise IdPs (ADFS, Okta, Ping), via `AssumeRoleWithSAML`; and **OIDC** for Cognito, Google, EKS service accounts, and GitHub Actions, via `AssumeRoleWithWebIdentity`. In both cases the external identity proves itself to the IdP, and AWS vends a short-lived role session in return.

The modern, preferred way to wire this across an organization is **IAM Identity Center** (formerly AWS SSO). It lives in your management account and connects to one source of truth — its own directory, Active Directory, or an external IdP like Okta or Entra ID. Users get **one sign-on portal** listing every account and role they can use; behind the scenes Identity Center calls `AssumeRole` and hands them temporary credentials.

Its unit of authorization is the **permission set** — a template bundling AWS-managed and customer-managed policies, an optional boundary, and an optional session duration. You define it once and assign it to users in specific accounts; Identity Center provisions a hidden IAM role into each target account, and updating the permission set updates that role everywhere it's assigned. **New AWS environments today should default to Identity Center over per-account IAM users** — it eliminates the whole category of long-lived console passwords and access keys.
