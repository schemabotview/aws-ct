## Networking an instance — SGs, IPs, ENIs

Four things wire an instance to the network. First, the **security group** — a **stateful, allow-only** firewall attached to the instance's network interface. Stateful means if you allow traffic in, the response is allowed back automatically; allow-only means there are no deny rules (for denies you drop to a subnet **NACL**). The default is deny-all inbound.

Second, **IP addresses**. A **private IP** is VPC-only, free, and permanent for the interface's life. A **public IP** is internet-routable, auto-assigned at launch, but **changes on stop/start** — and since 2024 every public IPv4 address is billed hourly. An **Elastic IP** is a public IP you own that **survives stop/start** — use it for a stable inbound endpoint or IP whitelisting.

Third, the **ENI** (Elastic Network Interface) — the virtual NIC every instance has (`eth0`), carrying the private IP, MAC, and security groups. **ENA** is the high-throughput driver (100+ Gbps, default on Nitro); **EFA** is a specialised ENI that bypasses the kernel for HPC and distributed-ML traffic.

Fourth, the **Instance Metadata Service** at `169.254.169.254`, where code discovers facts about its own instance — including **IAM role credentials**. Always require **IMDSv2** (token-based): IMDSv1's simple GET is vulnerable to SSRF and was the single biggest cause of historic EC2 credential theft.
